TASK 2: Cuisine Combination

```python
# Identify the most common combinations of cuisines in the dataset?

df_clean = df.dropna(subset=['Cuisines'])

def get_canonical_cuisine_combination(cuisine_str):

    cuisines = [c.strip() for c in cuisine_str.split(',')]

    cuisines.sort()

    return ', '.join(cuisines)


df_clean['Canonical Cuisines'] = df_clean['Cuisines'].apply(get_canonical_cuisine_combination)


top_combinations = df_clean['Canonical Cuisines'].value_counts().head(10).reset_index()
top_combinations.columns = ['Cuisine Combination', 'Count']


print("Top 10 Most Common Cuisine Combinations:")
print(top_combinations.to_markdown(index=False, numalign="left", stralign="left"))
```

```python
# Determine if certain cuisine combinations tend to have higher ratings?

df_clean = df.dropna(subset=['Cuisines'])

def get_canonical_cuisine_combination(cuisine_str):

    cuisines = [c.strip() for c in cuisine_str.split(',')]

    cuisines.sort()

    return ', '.join(cuisines)


df_clean['Canonical Cuisines'] = df_clean['Cuisines'].apply(get_canonical_cuisine_combination)


rating_by_cuisine = df_clean.groupby('Canonical Cuisines').agg(
    Count=('Aggregate rating', 'count'),
    Average_Rating=('Aggregate rating', 'mean')
).reset_index()


top_rated_combinations = rating_by_cuisine.sort_values(by='Count', ascending=False).head(10)


top_rated_combinations = top_rated_combinations.sort_values(by='Average_Rating', ascending=False)


top_rated_combinations['Average_Rating'] = top_rated_combinations['Average_Rating'].round(2)

top_rated_combinations = top_rated_combinations[['Canonical Cuisines', 'Average_Rating', 'Count']]
top_rated_combinations.columns = ['Cuisine Combination', 'Average Rating', 'Count']

print("Average Rating for Top 10 Most Frequent Cuisine Combinations (Sorted by Rating):")
print(top_rated_combinations.to_markdown(index=False, numalign="left", stralign="left"))
```

```python
# Analyze the ratings and popularity of different restaurant chains

restaurant_counts = df['Restaurant Name'].value_counts()
chain_names = restaurant_counts[restaurant_counts > 1].index
df_chains = df[df['Restaurant Name'].isin(chain_names)]

chain_analysis = df_chains.groupby('Restaurant Name').agg(
    Outlets=('Restaurant ID', 'count'),
    Average_Rating=('Aggregate rating', 'mean'),
    Total_Votes=('Votes', 'sum')
).reset_index()

top_10_largest_chains = chain_analysis.sort_values(by='Outlets', ascending=False).head(10)

well_established_chains = chain_analysis[chain_analysis['Outlets'] >= 5]
top_10_rated_chains = well_established_chains.sort_values(by='Average_Rating', ascending=False).head(10)
```

#Level 3

TASK 1:Restaurant Reviews

```python
#Analyze the text reviews to identify the most common positive and negative keywords.
#Define sentiment categories based on 'Rating text'
def categorize_sentiment(rating_text):
    if rating_text in ['Excellent', 'Very Good', 'Good']:
        return 'Positive'
    elif rating_text in ['Poor', 'Bad']:
        return 'Negative'
    elif rating_text in ['Average', 'Not rated']:
        return 'Neutral / Not Rated'
    return 'Other'


df['Sentiment Category'] = df['Rating text'].apply(categorize_sentiment)

sentiment_distribution = df['Sentiment Category'].value_counts(normalize=True).mul(100).round(2).reset_index()
sentiment_distribution.columns = ['Sentiment Category', 'Percentage']

print("--- WARNING: The dataset lacks a free-text review column. ---")
print("Analysis performed on the categorical 'Rating Text' labels as a proxy for sentiment.")
print("\nDistribution of Restaurant Sentiment Categories:")
print(sentiment_distribution.to_markdown(index=False, numalign="left", stralign="left"))
```

```python
#Calculate the average length of reviews and explore if there is a relationship between review length and rating

df['Rating Text Length'] = df['Rating text'].apply(len)
average_text_length = df['Rating Text Length'].mean()

relationship_analysis = df.groupby('Rating Text Length')['Aggregate rating'].mean().reset_index()

relationship_analysis.columns = ['Rating Text Length (Characters)', 'Average Aggregate Rating']
relationship_analysis['Average Aggregate Rating'] = relationship_analysis['Average Aggregate Rating'].round(2)

relationship_analysis = relationship_analysis.sort_values(by='Rating Text Length (Characters)', ascending=True)
```

TASK 3: Votes Analysis

```python
# Identify the restaurants with the highest and lowest number of votes.

max_votes = df['Votes'].max()
highest_voted_restaurants = df[df['Votes'] == max_votes]

min_votes = df['Votes'].min()
lowest_voted_restaurants = df[df['Votes'] == min_votes]

columns_to_display = ['Restaurant Name', 'City', 'Aggregate rating', 'Votes']

print("--- Restaurants with the HIGHEST Number of Votes ---")
print(f"Maximum Votes Recorded: {max_votes}")
print(highest_voted_restaurants[columns_to_display].to_markdown(index=False, numalign="left", stralign="left"))

print("\n--- Restaurants with the LOWEST Number of Votes ---")
print(f"Minimum Votes Recorded: {min_votes}")
print(lowest_voted_restaurants[columns_to_display].head(5).to_markdown(index=False, numalign="left", stralign="left"))
print(f"... and {len(lowest_voted_restaurants) - 5} more restaurants with 0 votes.")
```

```python
#Analyze if there is a correlation between the number of votes and the rating of a restaurant.

correlation = df['Votes'].corr(df['Aggregate rating'])
print(f"Pearson Correlation Coefficient between 'Votes' and 'Aggregate rating': {correlation:.4f}")

plt.figure(figsize=(10, 6))
plt.scatter(df['Votes'], df['Aggregate rating'], s=5, alpha=0.3, color='coral')
plt.title('Relationship between Number of Votes and Aggregate Rating')
plt.xlabel('Number of Votes')
plt.ylabel('Aggregate Rating')
plt.text(0.7, 0.9, f'Correlation (r): {correlation:.4f}', transform=plt.gca().transAxes, fontsize=12,
         verticalalignment='top', bbox=dict(boxstyle="round,pad=0.5", fc="white", alpha=0.7))
plt.yticks(range(0, 6))
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.savefig('votes_vs_rating_correlation.png')
```

TASK 3:Price Range vs. Online Delivery and
 Table Booking

```python
#Analyze if there is a relationship between the price range and the availability of online delivery and table booking.

# --- Analysis 1: Relationship between Price Range and Online Delivery ---
online_delivery_analysis = df.groupby('Price range')['Has Online delivery'].value_counts(normalize=True).mul(100).unstack(fill_value=0)
online_delivery_analysis.columns.name = None

# --- Analysis 2: Relationship between Price Range and Table Booking ---
table_booking_analysis = df.groupby('Price range')['Has Table booking'].value_counts(normalize=True).mul(100).unstack(fill_value=0)
table_booking_analysis.columns.name = None

online_delivery_analysis.index.name = 'Price Range'
table_booking_analysis.index.name = 'Price Range'

online_delivery_analysis = online_delivery_analysis.round(2)
table_booking_analysis = table_booking_analysis.round(2)

print("--- Relationship between Price Range and Availability of Online Delivery (Percentage) ---")
print(online_delivery_analysis.to_markdown(numalign="left", stralign="left"))

print("\n--- Relationship between Price Range and Availability of Table Booking (Percentage) ---")
print(table_booking_analysis.to_markdown(numalign="left", stralign="left"))
```

```python
#Determine if higher-priced restaurants are more likely to offer these services.

top_5_countries = df['Country Code'].value_counts().head(5).index.tolist()
df_top_countries = df[df['Country Code'].isin(top_5_countries)]
price_range_by_country = df_top_countries.groupby(['Country Code', 'Price range']).size().reset_index(name='Count')

plt.figure(figsize=(10, 6))
sns.barplot(x='Price range', y='Count', hue='Country Code', data=price_range_by_country, palette='tab10')
plt.title('Distribution of Price Ranges Across Top 5 Countries')
plt.xlabel('Price Range (1: Cheapest to 4: Most Expensive)')
plt.ylabel('Number of Restaurants')
plt.xticks(rotation=0)
plt.legend(title='Country Code')
plt.tight_layout()
plt.savefig('price_range_distribution_by_country.png')
print("--- Restaurant Count by Country Code and Price Range ---")
print(price_range_by_country.to_markdown(index=False, numalign="left", stralign="left"))
```
