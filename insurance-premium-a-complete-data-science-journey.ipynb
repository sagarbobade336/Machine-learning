### 2.3 Mathematical Foundation


üìê MATHEMATICAL FOUNDATION

The goal is to find a function f that maps input features X to target y:
    y = f(X) + Œµ
    
Where:
- y: Insurance charges (target variable)
- X: Customer features (age, BMI, smoker status, etc.)
- f: Unknown function we want to learn
- Œµ: Random error/noise

Our objective: Minimize the prediction error
    
Common Loss Functions for Regression:
1. MSE (Mean Squared Error) = (1/n) Œ£(yi - ≈∑i)¬≤
2. RMSE (Root MSE) = ‚àöMSE
3. MAE (Mean Absolute Error) = (1/n) Œ£|yi - ≈∑i|
4. R¬≤ (Coefficient of Determination) = 1 - (SS_res/SS_tot)



```python

# Demonstrate with a simple example
np.random.seed(42)
n_samples = 100

# Generate synthetic example
true_function = lambda x: 2*x + 3 + np.random.normal(0, 0.5, len(x))
x_demo = np.random.uniform(0, 10, n_samples)
y_demo = true_function(x_demo)

# Simple linear regression
from sklearn.linear_model import LinearRegression
model_demo = LinearRegression()
model_demo.fit(x_demo.reshape(-1, 1), y_demo)
y_pred_demo = model_demo.predict(x_demo.reshape(-1, 1))

# Visualization
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Plot 1: Data and fitted line
axes[0].scatter(x_demo, y_demo, alpha=0.5, label='Actual Data')
axes[0].plot(x_demo, y_pred_demo, 'r-', label='Fitted Line', linewidth=2)
axes[0].set_xlabel('Feature (X)')
axes[0].set_ylabel('Target (y)')
axes[0].set_title('Linear Regression Concept')
axes[0].legend()
axes[0].grid(True, alpha=0.3)

# Plot 2: Residuals
residuals = y_demo - y_pred_demo
axes[1].scatter(y_pred_demo, residuals, alpha=0.5)
axes[1].axhline(y=0, color='r', linestyle='--')
axes[1].set_xlabel('Predicted Values')
axes[1].set_ylabel('Residuals')
axes[1].set_title('Residual Plot (Should be randomly distributed)')
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# Calculate metrics
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error

mse = mean_squared_error(y_demo, y_pred_demo)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_demo, y_pred_demo)
r2 = r2_score(y_demo, y_pred_demo)

print("\nüìä METRICS EXPLAINED:")
print("="*50)
print(f"MSE  = {mse:.4f} ‚Üí Average squared error")
print(f"RMSE = {rmse:.4f} ‚Üí Error in same units as target")
print(f"MAE  = {mae:.4f} ‚Üí Average absolute error")
print(f"R¬≤   = {r2:.4f} ‚Üí Proportion of variance explained")
```

### 3.4 Key Business Metrics

Let me start by exploring the business fundamentals and then discover the key metrics that will shape our entire modeling strategy.

**The Economics of Health Insurance**

Health insurance operates on a fascinating principle called risk pooling. The idea is simple yet powerful: by spreading healthcare costs across a large group, we can make unpredictable individual expenses predictable and manageable at the group level. But here's where it gets interesting‚Äînot everyone contributes the same risk to the pool.
My first task is to understand just how different these risk levels can be. Let me start with what I suspect will be the most dramatic difference: 
**smoking status.**

```python
# I want to understand the real impact of smoking on healthcare costs
# Let's calculate this systematically

def analyze_smoking_impact(df):
    """
    My first deep dive: understanding how smoking affects insurance costs
    """
    # Group by smoking status and calculate comprehensive statistics
    smoking_analysis = df.groupby('smoker')['charges'].agg([
        'count', 
        'mean', 
        'median', 
        'std',
        'min',
        'max'
    ]).round(2)
    
    # Calculate the key metrics I need
    non_smoker_mean = smoking_analysis.loc['no', 'mean']
    smoker_mean = smoking_analysis.loc['yes', 'mean']
    multiplier = smoker_mean / non_smoker_mean
    percentage_increase = ((smoker_mean - non_smoker_mean) / non_smoker_mean) * 100
    
    print("üö¨ SMOKING IMPACT ANALYSIS")
    print("="*60)
    print(f"Non-smokers average cost: ${non_smoker_mean:,.2f}")
    print(f"Smokers average cost: ${smoker_mean:,.2f}")
    print(f"Cost multiplier: {multiplier:.2f}x")
    print(f"Percentage increase: {percentage_increase:.1f}%")
    print("\nDetailed Statistics:")
    print(smoking_analysis)
    
    # Store these for later reference
    return {
        'non_smoker_mean': non_smoker_mean,
        'smoker_mean': smoker_mean,
        'multiplier': multiplier,
        'percentage_increase': percentage_increase
    }

smoking_results = analyze_smoking_impact(df)
```

**Interpreting the Smoking Divide**

This is remarkable‚Äîsmokers cost 3.8 times more than non-smokers on average! We're talking about $32,050 versus $8,434 annually. This isn't just a statistical difference; it's a fundamental divide that explains why smoking status is often the first question on any insurance application.

But what does this mean for the business? If I were an insurance executive seeing this, I'd immediately think about two strategies: First, pricing must reflect this massive difference to remain profitable. Second, investing in smoking cessation programs could yield enormous returns. Even helping 10% of smoking customers quit could save millions in claims.

Now let me explore how age affects costs, which I expect to be more gradual but equally important.

```python
def analyze_age_cost_relationship(df):
    """
    Discovering the age-cost relationship: is it linear, exponential, or something else?
    """
    # Create age decades for clearer analysis
    df_age = df.copy()
    df_age['age_decade'] = (df_age['age'] // 10) * 10
    
    # First, let's use linear regression to find the average yearly increase
    from sklearn.linear_model import LinearRegression
    
    X = df_age[['age']].values
    y = df_age['charges'].values
    
    lr_model = LinearRegression()
    lr_model.fit(X, y)
    yearly_increase = lr_model.coef_[0]
    decade_increase = yearly_increase * 10
    
    # Now let's look at actual decade-by-decade progression
    decade_costs = df_age.groupby('age_decade')['charges'].mean()
    
    print("üìÖ AGE IMPACT ANALYSIS")
    print("="*60)
    print(f"Linear model findings:")
    print(f"  ‚Ä¢ Cost increase per year: ${yearly_increase:,.2f}")
    print(f"  ‚Ä¢ Cost increase per decade: ${decade_increase:,.2f}")
    
    print("\nActual average costs by age group:")
    for decade in sorted(decade_costs.index):
        cost = decade_costs[decade]
        count = len(df_age[df_age['age_decade'] == decade])
        print(f"  {decade}s: ${cost:,.2f} (n={count})")
    
    # Calculate the actual increases between decades
    print("\nDecade-over-decade increases:")
    for i, decade in enumerate(sorted(decade_costs.index)[1:], 1):
        prev_decade = sorted(decade_costs.index)[i-1]
        increase = decade_costs[decade] - decade_costs[prev_decade]
        pct_increase = (increase / decade_costs[prev_decade]) * 100
        print(f"  {prev_decade}s ‚Üí {decade}s: +${increase:,.2f} ({pct_increase:.1f}%)")
    
    return {
        'yearly_increase': yearly_increase,
        'decade_increase': decade_increase,
        'decade_costs': decade_costs
    }

age_results = analyze_age_cost_relationship(df)
```

**The Linear Age Progression**

Fascinating! Age shows a surprisingly linear relationship with costs, with each decade adding approximately $2,577 to annual charges. This is more predictable than I expected‚Äîmany insurance models assume exponential growth with age, but our data suggests a steadier progression, at least within the age range we're examining (18-64).

This linear relationship is actually good news for middle-aged customers who might be overcharged by exponential models. It also suggests that age-based pricing can be straightforward and transparent: "Each decade of life adds roughly $2,577 to your annual healthcare costs."

Let me now examine how BMI affects costs, which I suspect will show a threshold effect rather than a linear relationship.

```python
def analyze_bmi_impact(df):
    """
    Exploring the relationship between BMI and insurance costs
    Using WHO categories to make findings more actionable
    """
    df_bmi = df.copy()
    
    # Create BMI categories based on WHO standards
    # This mirrors how insurance companies actually think about weight
    df_bmi['bmi_category'] = pd.cut(
        df_bmi['bmi'],
        bins=[0, 18.5, 25, 30, 35, 100],
        labels=['Underweight', 'Normal', 'Overweight', 'Obese I', 'Obese II+']
    )
    
    # Calculate statistics for each category
    bmi_stats = df_bmi.groupby('bmi_category')['charges'].agg([
        'count', 'mean', 'median', 'std'
    ]).round(2)
    
    # Focus on the key business question: what's the cost of obesity?
    normal_mean = bmi_stats.loc['Normal', 'mean']
    
    # For business purposes, BMI > 30 is considered obese
    obese_data = df_bmi[df_bmi['bmi'] > 30]['charges']
    obese_mean = obese_data.mean()
    
    # Calculate the increase
    obese_increase = ((obese_mean - normal_mean) / normal_mean) * 100
    
    print("‚öñÔ∏è BMI IMPACT ANALYSIS")
    print("="*60)
    print(f"Key Finding:")
    print(f"  ‚Ä¢ Normal BMI (18.5-25) average cost: ${normal_mean:,.2f}")
    print(f"  ‚Ä¢ Obese (BMI > 30) average cost: ${obese_mean:,.2f}")
    print(f"  ‚Ä¢ Cost increase for obesity: {obese_increase:.1f}%")
    
    print("\nDetailed breakdown by BMI category:")
    print("-"*60)
    for category in bmi_stats.index:
        mean_cost = bmi_stats.loc[category, 'mean']
        count = bmi_stats.loc[category, 'count']
        pct_diff = ((mean_cost - normal_mean) / normal_mean) * 100
        sign = "+" if pct_diff > 0 else ""
        print(f"  {category:12s}: ${mean_cost:8,.2f} (n={count:3.0f}) [{sign}{pct_diff:6.1f}% vs normal]")
    
    # Also check if there's a linear correlation
    correlation = df_bmi['bmi'].corr(df_bmi['charges'])
    print(f"\nLinear correlation (BMI ‚Üî Charges): {correlation:.3f}")
    print("(Moderate correlation suggests non-linear relationship)")
    
    return {
        'normal_mean': normal_mean,
        'obese_mean': obese_mean,
        'increase_percentage': obese_increase,
        'bmi_stats': bmi_stats
    }

bmi_results = analyze_bmi_impact(df)
```

**BMI's Moderate but Significant Impact**

The 49.1% increase in costs for obesity (BMI > 30) is substantial but notably less dramatic than the smoking effect. This is an important finding for several reasons. First, it suggests that while weight is a legitimate risk factor, it shouldn't be weighted as heavily as smoking in pricing models. Second, the moderate increase makes a strong case for wellness programs rather than punitive pricing.

What's particularly interesting is the progression across categories‚Äîcosts don't jump suddenly at BMI 30 but increase gradually. This suggests that helping customers move down even one BMI category could yield meaningful cost savings.

Finally, let me examine regional variations, which could reflect differences in healthcare infrastructure, lifestyle, or local regulations.

```python
def analyze_regional_variations(df):
    """
    Exploring geographic cost variations
    Are these due to cost of living, lifestyle, or healthcare quality?
    """
    # Calculate comprehensive regional statistics
    regional_stats = df.groupby('region')['charges'].agg([
        'count', 'mean', 'median', 'std'
    ]).round(2)
    
    # Sort by mean cost to see the pattern clearly
    regional_stats = regional_stats.sort_values('mean')
    
    # Identify extremes
    lowest_region = regional_stats.index[0]
    highest_region = regional_stats.index[-1]
    lowest_cost = regional_stats.loc[lowest_region, 'mean']
    highest_cost = regional_stats.loc[highest_region, 'mean']
    
    # Calculate the spread
    regional_difference = ((highest_cost - lowest_cost) / lowest_cost) * 100
    
    print("üó∫Ô∏è REGIONAL VARIATION ANALYSIS")
    print("="*60)
    print(f"Regional cost ranking (lowest to highest):")
    print("-"*60)
    
    # Show all regions with relative differences
    baseline = regional_stats['mean'].min()
    for region in regional_stats.index:
        mean_cost = regional_stats.loc[region, 'mean']
        median_cost = regional_stats.loc[region, 'median']
        count = regional_stats.loc[region, 'count']
        diff_from_lowest = ((mean_cost - baseline) / baseline) * 100
        
        # Add a visual indicator for the difference
        bars = "‚ñà" * int(diff_from_lowest / 5)  # One bar per 5%
        
        print(f"  {region:12s}: ${mean_cost:8,.2f} (median: ${median_cost:8,.2f}, n={count:3.0f})")
        print(f"  {'':12s}  {bars} +{diff_from_lowest:.1f}%")
        print()
    
    print(f"Maximum regional difference: {regional_difference:.1f}%")
    print(f"This is relatively modest compared to smoking ({smoking_results['percentage_increase']:.0f}%)")
    
    return {
        'regional_stats': regional_stats,
        'max_difference': regional_difference,
        'lowest_region': lowest_region,
        'highest_region': highest_region
    }

regional_results = analyze_regional_variations(df)
```

**Regional Variations: Modest but Meaningful**

The 20% difference between Southeast and Northwest regions is significant enough to warrant attention but small enough to avoid major fairness concerns. This suggests that regional pricing adjustments should be modest‚Äîperhaps reflected in deductibles or network access rather than base premiums.

The Southeast's higher costs likely reflect a combination of factors: higher rates of chronic disease, different lifestyle patterns, and potentially less preventive care utilization. For national insurers, this suggests targeted wellness programs might be more effective than blanket pricing adjustments.

### 4.3 Advanced Statistical Analysis

Now let me apply more sophisticated statistical techniques to validate these observations and uncover deeper patterns.

```python
# Statistical tests and advanced analysis
print("\nüìä ADVANCED STATISTICAL ANALYSIS")
print("="*80)

# 1. Normality Tests
print("\n1. NORMALITY TESTS (Shapiro-Wilk):")
print("-"*50)
from scipy import stats

for col in df.select_dtypes(include=[np.number]).columns:
    if len(df[col]) > 5000:  # Shapiro-Wilk has sample size limitations
        stat, p_value = stats.normaltest(df[col])
        test_name = "D'Agostino-Pearson"
    else:
        stat, p_value = stats.shapiro(df[col])
        test_name = "Shapiro-Wilk"
    
    is_normal = "‚úì Normal" if p_value > 0.05 else "‚úó Not Normal"
    print(f"{col:10s}: p-value = {p_value:.6f} ‚Üí {is_normal} ({test_name})")

# 2. Outlier Detection
print("\n2. OUTLIER ANALYSIS (IQR Method):")
print("-"*50)

def detect_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
    return outliers, lower_bound, upper_bound

for col in ['age', 'bmi', 'charges']:
    outliers, lower, upper = detect_outliers_iqr(df, col)
    outlier_pct = len(outliers) / len(df) * 100
    print(f"{col:10s}: {len(outliers):3d} outliers ({outlier_pct:5.2f}%)")
    print(f"            Range: [{lower:.2f}, {upper:.2f}]")

# 3. ANOVA Tests
print("\n3. ANOVA TESTS (Categorical vs Charges):")
print("-"*50)

# One-way ANOVA for each categorical variable
categorical_vars = ['sex', 'smoker', 'region']

for cat_var in categorical_vars:
    groups = [group['charges'].values for name, group in df.groupby(cat_var)]
    f_stat, p_value = stats.f_oneway(*groups)
    
    effect = "Significant" if p_value < 0.05 else "Not Significant"
    print(f"{cat_var:10s}: F-stat = {f_stat:8.2f}, p-value = {p_value:.6f} ‚Üí {effect}")
    
    if p_value < 0.05:
        # Calculate effect size (eta squared)
        ss_between = sum(len(g) * (np.mean(g) - df['charges'].mean())**2 for g in groups)
        ss_total = sum((df['charges'] - df['charges'].mean())**2)
        eta_squared = ss_between / ss_total
        print(f"            Effect size (Œ∑¬≤) = {eta_squared:.4f}")

# 4. Chi-Square Tests for Independence
print("\n4. CHI-SQUARE TESTS (Categorical Independence):")
print("-"*50)

from scipy.stats import chi2_contingency

# Test independence between categorical variables
cat_pairs = [('sex', 'smoker'), ('sex', 'region'), ('smoker', 'region')]

for var1, var2 in cat_pairs:
    contingency_table = pd.crosstab(df[var1], df[var2])
    chi2, p_value, dof, expected = chi2_contingency(contingency_table)
    
    independent = "Independent" if p_value > 0.05 else "Dependent"
    print(f"{var1:8s} vs {var2:8s}: œá¬≤ = {chi2:7.2f}, p = {p_value:.6f} ‚Üí {independent}")

# 5. Multicollinearity Check (VIF)
print("\n5. MULTICOLLINEARITY CHECK (VIF):")
print("-"*50)
from statsmodels.stats.outliers_influence import variance_inflation_factor

# Prepare data for VIF calculation
df_vif = df.copy()

# Convert categorical variables to numeric
df_vif['sex'] = df_vif['sex'].map({'male': 1, 'female': 0})
df_vif['smoker'] = df_vif['smoker'].map({'yes': 1, 'no': 0})

# Create dummy variables for region
df_vif = pd.get_dummies(df_vif, columns=['region'], drop_first=True)

# Drop the target variable
df_vif = df_vif.drop('charges', axis=1)

# IMPORTANT: Select only numeric columns and convert to float
df_vif = df_vif.select_dtypes(include=[np.number])
df_vif = df_vif.astype(float)

# Check if there are any missing values
if df_vif.isnull().any().any():
    print("Warning: Missing values detected. Filling with mean...")
    df_vif = df_vif.fillna(df_vif.mean())

# Calculate VIF
vif_data = pd.DataFrame()
vif_data["Feature"] = df_vif.columns
vif_values = []

for i in range(len(df_vif.columns)):
    try:
        vif = variance_inflation_factor(df_vif.values, i)
        vif_values.append(vif)
    except Exception as e:
        print(f"Error calculating VIF for {df_vif.columns[i]}: {e}")
        vif_values.append(np.nan)

vif_data["VIF"] = vif_values
vif_data = vif_data.dropna()  # Remove any failed calculations
vif_data = vif_data.sort_values('VIF', ascending=False)

print("Variable Inflation Factors:")
for _, row in vif_data.iterrows():
    status = "‚úì OK" if row['VIF'] < 5 else "‚ö† Check" if row['VIF'] < 10 else "‚úó High"
    print(f"{row['Feature']:20s}: VIF = {row['VIF']:6.2f} {status}")

print("\nInterpretation: VIF < 5 (No multicollinearity), 5-10 (Moderate), >10 (High)")
```

### 5.1 Domain-Driven Feature Engineering

Domain features translate actuarial wisdom into model inputs. Rather than treating variables as abstract numbers, I'm encoding how insurance professionals actually think about risk‚ÄîBMI categories following WHO standards, age brackets reflecting life stages, and composite risk scores that insurers would recognize.

```python
# 5.1 Domain-Driven Feature Engineering 
import pandas as pd
import numpy as np
from sklearn.model_selection import KFold

class AdvancedFeatureEngineer:
    """
    Feature engineering pipeline that prevents data leakage.
    Target encoding is fitted only on training data.
    """
    
    def __init__(self, verbose=True):
        self.verbose = verbose
        self.target_encoders = {}  # Will store encoders fitted on training data
        self.is_fitted = False
        
    def create_domain_features(self, df):
        """
        Create domain features that DON'T use the target variable
        """
        df_feat = df.copy()
        
        if self.verbose:
            print("üìä Creating Domain-Based Features...")
            print("-"*50)
        
        # 1. BMI Categories (Safe - no target information)
        df_feat['bmi_category'] = pd.cut(
            df_feat['bmi'],
            bins=[0, 18.5, 25, 30, 35, 40, 100],
            labels=['Underweight', 'Normal', 'Overweight', 
                   'Obese_I', 'Obese_II', 'Obese_III']
        )
        
        df_feat['bmi_category_num'] = df_feat['bmi_category'].map({
            'Underweight': 1, 'Normal': 0, 'Overweight': 2,
            'Obese_I': 3, 'Obese_II': 4, 'Obese_III': 5
        })
        
        # 2. Age Groups (Safe)
        df_feat['age_group'] = pd.cut(
            df_feat['age'],
            bins=[17, 25, 35, 45, 55, 65],
            labels=['Young', 'Young_Adult', 'Middle_Age', 'Senior', 'Elder']
        )
        
        # 3. Family Size (Safe)
        df_feat['family_size'] = df_feat['children'].apply(
            lambda x: 'single' if x == 0 else 'small' if x <= 2 else 'large'
        )
        
        # 4. Health Status Score (Safe - based only on input features)
        def calculate_health_status(row):
            """Calculate health score based only on input features"""
            score = 100
            
            # BMI impact
            if row['bmi'] < 18.5 or row['bmi'] > 30:
                score -= 20
            elif row['bmi'] > 25:
                score -= 10
                
            # Smoking impact
            if row['smoker'] == 'yes':
                score -= 30  # Reduced from 40 to avoid being too predictive
                
            # Age impact
            if row['age'] > 50:
                score -= 10
            elif row['age'] > 40:
                score -= 5
                
            return score
        
        df_feat['health_score'] = df_feat.apply(calculate_health_status, axis=1)
        
        # 5. Age-BMI interaction (Safe)
        df_feat['age_bmi_interaction'] = df_feat['age'] * df_feat['bmi'] / 100
        
        if self.verbose:
            print(f"‚úÖ Created {5} domain features")
            
        return df_feat
    
    def create_statistical_features(self, df):
        """
        Create statistical transformations (all safe)
        """
        df_feat = df.copy()
        
        if self.verbose:
            print("\nüìà Creating Statistical Features...")
            print("-"*50)
        
        # Log transformations
        df_feat['log_bmi'] = np.log1p(df_feat['bmi'])
        df_feat['log_age'] = np.log1p(df_feat['age'])
        
        # Square root transformations
        df_feat['sqrt_age'] = np.sqrt(df_feat['age'])
        df_feat['sqrt_bmi'] = np.sqrt(df_feat['bmi'])
        
        # Polynomial features (be careful not to overfit)
        df_feat['age_squared'] = df_feat['age'] ** 2
        df_feat['bmi_squared'] = df_feat['bmi'] ** 2
        
        # Binning
        df_feat['age_decade'] = (df_feat['age'] // 10) * 10
        
        if self.verbose:
            print(f"‚úÖ Created {7} statistical features")
            
        return df_feat
    
    def create_interaction_features(self, df):
        """
        Create interaction features between variables
        """
        df_feat = df.copy()
        
        if self.verbose:
            print("\nüîÑ Creating Interaction Features...")
            print("-"*50)
        
        # Age-BMI interactions
        df_feat['age_bmi_ratio'] = df_feat['age'] / (df_feat['bmi'] + 1)
        
        # Smoker interactions (binary encoding)
        smoker_binary = df_feat['smoker'].map({'yes': 1, 'no': 0})
        df_feat['smoker_age'] = smoker_binary * df_feat['age']
        df_feat['smoker_bmi'] = smoker_binary * df_feat['bmi']
        
        # Children interactions
        df_feat['children_per_age'] = df_feat['children'] / (df_feat['age'] + 1)
        
        # Sex interactions
        male_binary = (df_feat['sex'] == 'male').astype(int)
        df_feat['male_bmi'] = male_binary * df_feat['bmi']
        df_feat['male_age'] = male_binary * df_feat['age']
        
        if self.verbose:
            print(f"‚úÖ Created {6} interaction features")
            
        return df_feat
    
    def fit_target_encoding(self, X_train, y_train, cat_columns):
        """
        Fit target encoders on training data only
        This should be called AFTER train-test split
        """
        if self.verbose:
            print("\nüéØ Fitting Target Encoders on Training Data...")
            print("-"*50)
        
        self.target_encoders = {}
        
        for col in cat_columns:
            if col in X_train.columns:
                # Calculate mean target for each category
                encoding_dict = {}
                for category in X_train[col].unique():
                    mask = X_train[col] == category
                    encoding_dict[category] = y_train[mask].mean()
                
                # Store the global mean for unknown categories
                global_mean = y_train.mean()
                encoding_dict['_default'] = global_mean
                
                self.target_encoders[col] = encoding_dict
                
        self.is_fitted = True
        
        if self.verbose:
            print(f"‚úÖ Fitted target encoders for {len(self.target_encoders)} columns")
    
    def transform_target_encoding(self, X):
        """
        Apply target encoding using pre-fitted encoders
        """
        if not self.is_fitted:
            raise ValueError("Must call fit_target_encoding first!")
        
        X_encoded = X.copy()
        
        for col, encoding_dict in self.target_encoders.items():
            if col in X_encoded.columns:
                default_value = encoding_dict['_default']
                X_encoded[f'{col}_target_mean'] = X_encoded[col].map(
                    lambda x: encoding_dict.get(x, default_value)
                )
        
        return X_encoded
    
    def create_safe_features(self, df):
        """
        Create all features that don't require target variable
        """
        if self.verbose:
            print("\nüöÄ SAFE FEATURE ENGINEERING PIPELINE")
            print("="*60)
        
        df_feat = df.copy()
        df_feat = self.create_domain_features(df_feat)
        df_feat = self.create_statistical_features(df_feat)
        df_feat = self.create_interaction_features(df_feat)
        
        if self.verbose:
            print("\nüìä FEATURE ENGINEERING SUMMARY")
            print("="*60)
            print(f"Original features: {df.shape[1]}")
            print(f"New features created: {df_feat.shape[1] - df.shape[1]}")
            print(f"Total features: {df_feat.shape[1]}")
        
        return df_feat

# Usage example:
print("üîß Applying Safe Feature Engineering...")
print("="*80)

# Step 1: Create features that don't need target
engineer = AdvancedFeatureEngineer(verbose=True)
df_engineered = engineer.create_safe_features(df)

print("\n‚úÖ Safe feature engineering complete!")
print(f"Dataset shape: {df_engineered.shape}")
```

My domain-based features capture insurance-specific risk factors, particularly through the health score metric that combines BMI, smoking status, and age into a single risk indicator. This mirrors how actuaries actually assess insurance risk. The BMI categorization and age grouping follow standard medical classifications, ensuring my features align with industry practices.

### 5.2 Feature Selection and Importance

Creating features is only half the battle‚Äîidentifying which ones actually improve predictions is crucial. Using multiple selection methods (correlation, mutual information, tree importance, embedded methods), I'll determine which features carry real signal versus just adding noise.

```python
# 5.2 Feature Selection (CORRECTED - Must be done AFTER train-test split)
from sklearn.feature_selection import SelectKBest, f_regression, mutual_info_regression
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

def safe_feature_selection(X_train, y_train, top_k=20):
    """
    Perform feature selection ONLY on training data to prevent leakage
    Returns selected feature names and a transformer fitted on training data
    """
    print("\nüéØ FEATURE SELECTION ON TRAINING DATA ONLY")
    print("="*80)
    
    # Ensure we're working with copies
    X_train_select = X_train.copy()
    
    # Encode categorical variables for selection process
    label_encoders = {}
    for col in X_train_select.columns:
        if X_train_select[col].dtype == 'object' or X_train_select[col].dtype.name == 'category':
            le = LabelEncoder()
            X_train_select[col] = le.fit_transform(X_train_select[col].astype(str))
            label_encoders[col] = le
    
    # Handle NaN values
    for col in X_train_select.columns:
        if X_train_select[col].isnull().any():
            X_train_select[col] = X_train_select[col].fillna(X_train_select[col].median())
    
    # Ensure all columns are numeric
    X_train_select = X_train_select.apply(pd.to_numeric, errors='coerce').fillna(0)
    
    # 1. Correlation with target (using only training data)
    print("\n1. CORRELATION-BASED SELECTION")
    print("-"*50)
    correlations = X_train_select.corrwith(y_train).abs().sort_values(ascending=False)
    
    print("Top 10 features by correlation:")
    for i, (feat, corr) in enumerate(correlations.head(10).items(), 1):
        print(f"{i:2d}. {feat:25s}: {corr:.4f}")
    
    # 2. F-statistic (ANOVA) - fitted only on training data
    print("\n2. F-STATISTIC (ANOVA) SELECTION")
    print("-"*50)
    selector_f = SelectKBest(score_func=f_regression, k=min(top_k, X_train_select.shape[1]))
    selector_f.fit(X_train_select, y_train)
    f_scores = pd.Series(selector_f.scores_, index=X_train_select.columns).sort_values(ascending=False)
    
    print("Top 10 features by F-statistic:")
    for i, (feat, score) in enumerate(f_scores.head(10).items(), 1):
        print(f"{i:2d}. {feat:25s}: {score:10.2f}")
    
    # 3. Mutual Information - calculated only on training data
    print("\n3. MUTUAL INFORMATION SELECTION")
    print("-"*50)
    mi_scores = mutual_info_regression(X_train_select, y_train, random_state=42)
    mi_scores = pd.Series(mi_scores, index=X_train_select.columns).sort_values(ascending=False)
    
    print("Top 10 features by Mutual Information:")
    for i, (feat, score) in enumerate(mi_scores.head(10).items(), 1):
        print(f"{i:2d}. {feat:25s}: {score:.4f}")
    
    # 4. Random Forest Importance - fitted only on training data
    print("\n4. RANDOM FOREST IMPORTANCE")
    print("-"*50)
    rf = RandomForestRegressor(n_estimators=50, max_depth=5, random_state=42, n_jobs=-1)
    rf.fit(X_train_select, y_train)
    rf_importance = pd.Series(rf.feature_importances_, index=X_train_select.columns).sort_values(ascending=False)
    
    print("Top 10 features by Random Forest:")
    for i, (feat, imp) in enumerate(rf_importance.head(10).items(), 1):
        print(f"{i:2d}. {feat:25s}: {imp:.4f}")
    
    # 5. Ensemble approach
    print("\n5. ENSEMBLE FEATURE SELECTION")
    print("-"*50)
    
    # Normalize scores
    def safe_normalize(scores):
        max_val = scores.max()
        if max_val > 0:
            return scores / max_val
        return scores
    
    norm_corr = safe_normalize(correlations.fillna(0))
    norm_f = safe_normalize(f_scores.fillna(0))
    norm_mi = safe_normalize(mi_scores.fillna(0))
    norm_rf = safe_normalize(rf_importance.fillna(0))
    
    # Weighted ensemble (reduced Random Forest weight to avoid overfitting)
    ensemble_scores = (
        0.25 * norm_corr +
        0.25 * norm_f +
        0.30 * norm_mi +
        0.20 * norm_rf
    ).sort_values(ascending=False)
    
    print(f"Top {min(20, top_k)} features by Ensemble Score:")
    for i, (feat, score) in enumerate(ensemble_scores.head(min(20, top_k)).items(), 1):
        print(f"{i:2d}. {feat:25s}: {score:.4f}")
    
    # Visualization
    fig, axes = plt.subplots(2, 2, figsize=(15, 12))
    
    # Plot comparisons (same as before but with training data only)
    methods = [
        (correlations, 'Correlation', axes[0, 0]),
        (f_scores, 'ANOVA F-test', axes[0, 1]),
        (mi_scores, 'Mutual Information', axes[1, 0]),
        (ensemble_scores, 'Ensemble Method', axes[1, 1])
    ]
    
    for scores, title, ax in methods:
        top15 = scores.head(15)
        ax.barh(range(len(top15)), top15.values[::-1])
        ax.set_yticks(range(len(top15)))
        ax.set_yticklabels(top15.index[::-1])
        ax.set_xlabel('Score')
        ax.set_title(f'Top Features by {title}', fontweight='bold')
        ax.grid(True, alpha=0.3)
    
    plt.suptitle('Feature Selection Methods (Training Data Only)', fontsize=16, fontweight='bold')
    plt.tight_layout()
    plt.show()
    
    # Select top K features
    selected_features = ensemble_scores.head(top_k).index.tolist()
    
    print(f"\n‚úÖ Selected {len(selected_features)} features based on training data")
    
    return selected_features, ensemble_scores, label_encoders

# This function should be called AFTER train-test split
def apply_feature_selection_pipeline(df_engineered, target_col='charges', test_size=0.2, top_k=20):
    """
    Complete pipeline: split first, then select features
    """
    print("\nüîÑ CORRECT FEATURE SELECTION PIPELINE")
    print("="*80)
    
    # STEP 1: Split data FIRST
    print("\nStep 1: Splitting data into train and test sets...")
    X = df_engineered.drop(target_col, axis=1)
    y = df_engineered[target_col]
    
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=test_size, random_state=42
    )
    print(f"Training set: {X_train.shape}")
    print(f"Test set: {X_test.shape}")
    
    # STEP 2: Select features using ONLY training data
    print("\nStep 2: Selecting features on training data...")
    selected_features, feature_scores, label_encoders = safe_feature_selection(
        X_train, y_train, top_k=top_k
    )
    
    # STEP 3: Apply feature selection to both sets
    print("\nStep 3: Applying selected features to both sets...")
    X_train_selected = X_train[selected_features]
    X_test_selected = X_test[selected_features]
    
    print(f"\n‚úÖ Final shapes after feature selection:")
    print(f"X_train: {X_train_selected.shape}")
    print(f"X_test: {X_test_selected.shape}")
    
    return {
        'X_train': X_train_selected,
        'X_test': X_test_selected,
        'y_train': y_train,
        'y_test': y_test,
        'selected_features': selected_features,
        'feature_scores': feature_scores,
        'label_encoders': label_encoders
    }

# Execute the corrected pipeline
selection_results = apply_feature_selection_pipeline(
    df_engineered, 
    target_col='charges',
    test_size=0.2,
    top_k=20  # Reduced from 30 to prevent overfitting
)
```

I successfully implemented a robust feature selection pipeline that properly prevents data leakage by performing the selection exclusively on training data. After correctly splitting my dataset (1070 training samples, 268 test samples), I applied four complementary selection methods to identify the most predictive features from my 26 engineered features.

### 6.1 Data Preparation Pipeline

Before any model can learn, we need a robust, reproducible pipeline that transforms our engineered features into model-ready inputs. This isn't glamorous work, but it's where many real-world projects fail‚Äîthrough data leakage, improper scaling, or irreproducible train-test splits. I'm building a pipeline that handles these challenges systematically.

```python
# 6.1 Data Preparation Pipeline (CORRECTED)
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, RobustScaler, LabelEncoder
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
import pandas as pd
import numpy as np

class SafeDataPreparationPipeline:
    """
    Data preparation pipeline that prevents data leakage by:
    1. Splitting data FIRST
    2. Fitting encoders and scalers ONLY on training data
    3. Applying target encoding ONLY after split
    """
    
    def __init__(self, test_size=0.2, val_size=0.2, random_state=42):
        self.test_size = test_size
        self.val_size = val_size
        self.random_state = random_state
        self.label_encoders = {}
        self.scaler = None
        self.feature_columns = None
        self.target_encoders = {}
        
    def prepare_data(self, df, target_col='charges', selected_features=None, engineer=None):
        """
        Properly prepare data with correct order of operations
        """
        print("\nüìä SAFE DATA PREPARATION PIPELINE")
        print("="*80)
        
        # STEP 1: Separate features and target
        print("\n1. Separating features and target...")
        if selected_features:
            # Use only selected features
            feature_cols = [col for col in selected_features if col in df.columns and col != target_col]
            X = df[feature_cols].copy()
        else:
            X = df.drop(target_col, axis=1).copy()
        
        y = df[target_col].copy()
        
        print(f"   ‚úì Features shape: {X.shape}")
        print(f"   ‚úì Target shape: {y.shape}")
        
        # STEP 2: Split data BEFORE any fitting
        print("\n2. Splitting data (BEFORE any fitting)...")
        X_train, X_test, y_train, y_test = train_test_split(
            X, y, test_size=self.test_size, random_state=self.random_state, stratify=None
        )
        
        # Further split training into train and validation
        X_train, X_val, y_train, y_val = train_test_split(
            X_train, y_train, test_size=self.val_size, random_state=self.random_state
        )
        
        print(f"   ‚úì Training set: {X_train.shape[0]} samples")
        print(f"   ‚úì Validation set: {X_val.shape[0]} samples")
        print(f"   ‚úì Test set: {X_test.shape[0]} samples")
        
        # STEP 3: Fit label encoders on training data only
        print("\n3. Encoding categorical variables (fitted on training only)...")
        
        # Find categorical columns
        categorical_columns = []
        for col in X_train.columns:
            if X_train[col].dtype == 'object' or X_train[col].dtype.name == 'category':
                categorical_columns.append(col)
        
        # Fit encoders on training data
        for col in categorical_columns:
            le = LabelEncoder()
            # Fit on training data
            le.fit(X_train[col].astype(str))
            self.label_encoders[col] = le
            
            # Transform all sets
            X_train[col] = le.transform(X_train[col].astype(str))
            X_val[col] = X_val[col].astype(str).map(
                lambda x: le.transform([x])[0] if x in le.classes_ else -1
            )
            X_test[col] = X_test[col].astype(str).map(
                lambda x: le.transform([x])[0] if x in le.classes_ else -1
            )
            
            print(f"   ‚úì Encoded {col}: {len(le.classes_)} categories")
        
        # STEP 4: Handle missing values (using training statistics)
        print("\n4. Handling missing values (using training statistics)...")
        
        numeric_columns = X_train.select_dtypes(include=[np.number]).columns
        
        for col in numeric_columns:
            # Calculate median from training data
            train_median = X_train[col].median()
            
            # Apply to all sets
            X_train[col] = X_train[col].fillna(train_median)
            X_val[col] = X_val[col].fillna(train_median)
            X_test[col] = X_test[col].fillna(train_median)
        
        # STEP 5: Apply target encoding (if engineer provided)
        if engineer is not None and hasattr(engineer, 'fit_target_encoding'):
            print("\n5. Applying target encoding (fitted on training only)...")
            
            # Fit on training data
            cat_cols = ['sex', 'smoker', 'region', 'family_size', 'age_group']
            available_cats = [col for col in cat_cols if col in X_train.columns]
            
            if available_cats:
                engineer.fit_target_encoding(X_train, y_train, available_cats)
                
                # Transform all sets
                X_train = engineer.transform_target_encoding(X_train)
                X_val = engineer.transform_target_encoding(X_val)
                X_test = engineer.transform_target_encoding(X_test)
                
                print(f"   ‚úì Added target encoding for {len(available_cats)} columns")
        
        # STEP 6: Scale features (fit on training only)
        print("\n6. Scaling features (fitted on training only)...")
        
        # Ensure all data is numeric
        X_train = X_train.apply(pd.to_numeric, errors='coerce').fillna(0)
        X_val = X_val.apply(pd.to_numeric, errors='coerce').fillna(0)
        X_test = X_test.apply(pd.to_numeric, errors='coerce').fillna(0)
        
        # Fit scaler on training data
        self.scaler = RobustScaler()
        X_train_scaled = self.scaler.fit_transform(X_train)
        X_val_scaled = self.scaler.transform(X_val)
        X_test_scaled = self.scaler.transform(X_test)
        
        # Convert back to DataFrame
        X_train_scaled = pd.DataFrame(X_train_scaled, columns=X_train.columns, index=X_train.index)
        X_val_scaled = pd.DataFrame(X_val_scaled, columns=X_val.columns, index=X_val.index)
        X_test_scaled = pd.DataFrame(X_test_scaled, columns=X_test.columns, index=X_test.index)
        
        print("   ‚úì Applied RobustScaler")
        
        # STEP 7: Final statistics
        print("\n7. Final Data Statistics:")
        print("-"*50)
        print(f"   Target (training) range: ${y_train.min():,.2f} - ${y_train.max():,.2f}")
        print(f"   Target (training) mean: ${y_train.mean():,.2f}")
        print(f"   Target (training) std: ${y_train.std():,.2f}")
        
        # Check for potential issues
        print("\n8. Data Quality Checks:")
        print("-"*50)
        
        # Check for NaN values
        train_nans = X_train_scaled.isnull().sum().sum()
        val_nans = X_val_scaled.isnull().sum().sum()
        test_nans = X_test_scaled.isnull().sum().sum()
        
        print(f"   NaN values - Train: {train_nans}, Val: {val_nans}, Test: {test_nans}")
        
        # Check for infinite values
        train_inf = np.isinf(X_train_scaled).sum().sum()
        val_inf = np.isinf(X_val_scaled).sum().sum()
        test_inf = np.isinf(X_test_scaled).sum().sum()
        
        print(f"   Infinite values - Train: {train_inf}, Val: {val_inf}, Test: {test_inf}")
        
        # Feature variance check (on training data)
        low_variance = (X_train_scaled.var() < 0.01).sum()
        print(f"   Low variance features (<0.01): {low_variance}")
        
        print("\n‚úÖ Data preparation complete (no data leakage)!")
        
        return {
            'X_train': X_train_scaled,
            'X_val': X_val_scaled,
            'X_test': X_test_scaled,
            'y_train': y_train,
            'y_val': y_val,
            'y_test': y_test,
            'X_train_unscaled': X_train,
            'X_val_unscaled': X_val,
            'X_test_unscaled': X_test
        }

# Execute the corrected pipeline
print("üîß Executing Safe Data Preparation Pipeline...")
print("="*80)

# Use the corrected feature engineer from earlier
prep_pipeline = SafeDataPreparationPipeline(test_size=0.2, val_size=0.2, random_state=42)

# If you're using selected features from the corrected feature selection:
if 'selection_results' in locals():
    # Reconstruct the dataset with selected features
    df_for_modeling = df_engineered[selection_results['selected_features'] + ['charges']]
    data = prep_pipeline.prepare_data(
        df_for_modeling,
        target_col='charges',
        selected_features=selection_results['selected_features'],
        engineer=engineer if 'engineer' in locals() else None
    )
else:
    # Use all features
    data = prep_pipeline.prepare_data(
        df_engineered,
        target_col='charges',
        engineer=engineer if 'engineer' in locals() else None
    )

# Extract prepared data
X_train = data['X_train']
X_val = data['X_val']
X_test = data['X_test']
y_train = data['y_train']
y_val = data['y_val']
y_test = data['y_test']

print("\nüìä Final Data Shapes:")
print("-"*40)
print(f"X_train: {X_train.shape}")
print(f"X_val:   {X_val.shape}")
print(f"X_test:  {X_test.shape}")
print(f"y_train: {y_train.shape}")
print(f"y_val:   {y_val.shape}")
print(f"y_test:  {y_test.shape}")
```

I've successfully implemented a robust data preparation pipeline that strictly prevents data leakage by following the correct order of operations. Starting with my 20 selected features from the previous feature selection step, I processed 1,338 total samples through a carefully sequenced pipeline.

My data splits follow a 64-16-20 ratio, yielding 856 training samples, 214 validation samples, and 268 test samples. ### 6.3 Advanced Models - XGBoost, LightGBM, CatBoost

Given Gradient Boosting's exceptional baseline performance, advanced gradient boosting implementations should push us even closer to the theoretical performance limit. These algorithms‚ÄîXGBoost, LightGBM, CatBoost‚Äîoffer sophisticated regularization and optimization techniques that could extract the last bits of predictive power from our features.

```python
# 6.3 Advanced Models (CORRECTED with strong regularization)
import xgboost as xgb
import lightgbm as lgb
from catboost import CatBoostRegressor
import optuna
from optuna.samplers import TPESampler
import warnings
warnings.filterwarnings('ignore')

class SafeAdvancedModels:
    """
    Train advanced models with STRONG regularization to prevent overfitting
    Key changes:
    - Reduced model complexity
    - Added early stopping
    - More conservative hyperparameter ranges
    - Cross-validation during optimization
    """
    
    def __init__(self, use_optuna=False, n_trials=20):  # Reduced trials, optuna off by default
        self.use_optuna = use_optuna
        self.n_trials = n_trials
        self.models = {}
        self.best_params = {}
        self.results = {}
        
    def optimize_xgboost(self, X_train, y_train, X_val, y_val):
        """
        XGBoost with conservative parameters to prevent overfitting
        """
        print("\nüéØ Training XGBoost with Regularization...")
        print("-"*50)
        
        if self.use_optuna:
            def objective(trial):
                params = {
                    'n_estimators': trial.suggest_int('n_estimators', 50, 200),  # Reduced range
                    'max_depth': trial.suggest_int('max_depth', 2, 5),  # Much shallower
                    'learning_rate': trial.suggest_float('learning_rate', 0.01, 0.1, log=True),
                    'subsample': trial.suggest_float('subsample', 0.5, 0.8),  # More conservative
                    'colsample_bytree': trial.suggest_float('colsample_bytree', 0.5, 0.8),
                    'gamma': trial.suggest_float('gamma', 0.1, 1.0),  # Increased minimum
                    'reg_alpha': trial.suggest_float('reg_alpha', 0.1, 5.0),  # L1 regularization
                    'reg_lambda': trial.suggest_float('reg_lambda', 1.0, 10.0),  # L2 regularization
                    'min_child_weight': trial.suggest_int('min_child_weight', 5, 20),  # Increased
                    'random_state': 42
                }
                
                # Use cross-validation instead of single validation set
                model = xgb.XGBRegressor(**params)
                cv_scores = cross_val_score(
                    model, X_train, y_train,
                    cv=3,  # 3-fold CV
                    scoring='neg_root_mean_squared_error'
                )
                return -cv_scores.mean()
            
            study = optuna.create_study(
                direction='minimize',
                sampler=TPESampler(seed=42)
            )
            optuna.logging.set_verbosity(optuna.logging.WARNING)
            study.optimize(objective, n_trials=self.n_trials, show_progress_bar=False)
            
            self.best_params['xgboost'] = study.best_params
            print(f"‚úÖ Best CV RMSE: ${study.best_value:,.2f}")
            
            # Train final model with best params
            model = xgb.XGBRegressor(**study.best_params, random_state=42)
        else:
            # Use conservative default parameters
            model = xgb.XGBRegressor(
                n_estimators=100,
                max_depth=3,  # Shallow trees
                learning_rate=0.05,
                subsample=0.7,
                colsample_bytree=0.7,
                gamma=0.1,  # Minimum loss reduction
                reg_alpha=1.0,  # L1 regularization
                reg_lambda=5.0,  # L2 regularization
                min_child_weight=10,
                random_state=42
            )
        
        # Train with early stopping
        model.fit(
            X_train, y_train,
            eval_set=[(X_val, y_val)],
            early_stopping_rounds=20,  # Reduced from 50
            verbose=False
        )
        
        self.models['xgboost'] = model
        return model
    
    def optimize_lightgbm(self, X_train, y_train, X_val, y_val):
        """
        LightGBM with conservative parameters
        """
        print("\nüéØ Training LightGBM with Regularization...")
        print("-"*50)
        
        if self.use_optuna:
            def objective(trial):
                params = {
                    'n_estimators': trial.suggest_int('n_estimators', 50, 200),
                    'max_depth': trial.suggest_int('max_depth', 2, 5),
                    'learning_rate': trial.suggest_float('learning_rate', 0.01, 0.1, log=True),
                    'num_leaves': trial.suggest_int('num_leaves', 10, 50),  # Reduced
                    'min_child_samples': trial.suggest_int('min_child_samples', 20, 50),  # Increased
                    'subsample': trial.suggest_float('subsample', 0.5, 0.8),
                    'colsample_bytree': trial.suggest_float('colsample_bytree', 0.5, 0.8),
                    'reg_alpha': trial.suggest_float('reg_alpha', 0.1, 5.0),
                    'reg_lambda': trial.suggest_float('reg_lambda', 1.0, 10.0),
                    'min_split_gain': trial.suggest_float('min_split_gain', 0.01, 0.1),
                    'random_state': 42,
                    'verbose': -1
                }
                
                model = lgb.LGBMRegressor(**params)
                cv_scores = cross_val_score(
                    model, X_train, y_train,
                    cv=3,
                    scoring='neg_root_mean_squared_error'
                )
                return -cv_scores.mean()
            
            study = optuna.create_study(
                direction='minimize',
                sampler=TPESampler(seed=42)
            )
            optuna.logging.set_verbosity(optuna.logging.WARNING)
            study.optimize(objective, n_trials=self.n_trials, show_progress_bar=False)
            
            self.best_params['lightgbm'] = study.best_params
            print(f"‚úÖ Best CV RMSE: ${study.best_value:,.2f}")
            
            model = lgb.LGBMRegressor(**study.best_params, random_state=42, verbose=-1)
        else:
            model = lgb.LGBMRegressor(
                n_estimators=100,
                max_depth=3,
                learning_rate=0.05,
                num_leaves=20,
                min_child_samples=30,
                subsample=0.7,
                colsample_bytree=0.7,
                reg_alpha=1.0,
                reg_lambda=5.0,
                min_split_gain=0.01,
                random_state=42,
                verbose=-1
            )
        
        model.fit(
            X_train, y_train,
            eval_set=[(X_val, y_val)],
            callbacks=[lgb.early_stopping(20), lgb.log_evaluation(0)]
        )
        
        self.models['lightgbm'] = model
        return model
    
    def optimize_catboost(self, X_train, y_train, X_val, y_val):
        """
        CatBoost with conservative parameters
        """
        print("\nüéØ Training CatBoost with Regularization...")
        print("-"*50)
        
        if self.use_optuna:
            def objective(trial):
                params = {
                    'iterations': trial.suggest_int('iterations', 50, 200),
                    'depth': trial.suggest_int('depth', 2, 5),
                    'learning_rate': trial.suggest_float('learning_rate', 0.01, 0.1, log=True),
                    'l2_leaf_reg': trial.suggest_float('l2_leaf_reg', 3, 10),  # Increased
                    'border_count': trial.suggest_int('border_count', 32, 128),
                    'bagging_temperature': trial.suggest_float('bagging_temperature', 0.5, 1),
                    'random_strength': trial.suggest_float('random_strength', 0.5, 1),
                    'min_data_in_leaf': trial.suggest_int('min_data_in_leaf', 10, 30),  # Added
                    'random_state': 42,
                    'verbose': False
                }
                
                model = CatBoostRegressor(**params)
                cv_scores = cross_val_score(
                    model, X_train, y_train,
                    cv=3,
                    scoring='neg_root_mean_squared_error'
                )
                return -cv_scores.mean()
            
            study = optuna.create_study(
                direction='minimize',
                sampler=TPESampler(seed=42)
            )
            optuna.logging.set_verbosity(optuna.logging.WARNING)
            study.optimize(objective, n_trials=self.n_trials, show_progress_bar=False)
            
            self.best_params['catboost'] = study.best_params
            print(f"‚úÖ Best CV RMSE: ${study.best_value:,.2f}")
            
            model = CatBoostRegressor(**study.best_params, random_state=42, verbose=False)
        else:
            model = CatBoostRegressor(
                iterations=100,
                depth=3,
                learning_rate=0.05,
                l2_leaf_reg=5,
                min_data_in_leaf=20,
                bagging_temperature=0.7,
                random_strength=0.7,
                random_state=42,
                verbose=False
            )
        
        model.fit(
            X_train, y_train,
            eval_set=(X_val, y_val),
            early_stopping_rounds=20,
            verbose=False
        )
        
        self.models['catboost'] = model
        return model
    
    def train_all_advanced_models(self, X_train, y_train, X_val, y_val, X_test=None, y_test=None):
        """
        Train all advanced models with overfitting checks
        """
        print("\nüöÄ TRAINING REGULARIZED ADVANCED MODELS")
        print("="*80)
        
        # Train models
        self.optimize_xgboost(X_train, y_train, X_val, y_val)
        self.optimize_lightgbm(X_train, y_train, X_val, y_val)
        self.optimize_catboost(X_train, y_train, X_val, y_val)
        
        # Evaluate all models
        print("\nüìä ADVANCED MODELS PERFORMANCE (WITH OVERFITTING CHECK)")
        print("="*80)
        
        for name, model in self.models.items():
            # Predictions
            y_train_pred = model.predict(X_train)
            y_val_pred = model.predict(X_val)
            
            # Metrics
            train_rmse = np.sqrt(mean_squared_error(y_train, y_train_pred))
            val_rmse = np.sqrt(mean_squared_error(y_val, y_val_pred))
            train_r2 = r2_score(y_train, y_train_pred)
            val_r2 = r2_score(y_val, y_val_pred)
            train_mae = mean_absolute_error(y_train, y_train_pred)
            val_mae = mean_absolute_error(y_val, y_val_pred)
            
            # Cross-validation for robustness
            cv_scores = cross_val_score(
                model, X_train, y_train,
                cv=5, scoring='neg_root_mean_squared_error'
            )
            cv_rmse = -cv_scores.mean()
            cv_std = cv_scores.std()
            
            # Calculate overfitting metrics
            overfitting_gap = train_r2 - val_r2
            
            # Test set performance (if provided)
            if X_test is not None and y_test is not None:
                y_test_pred = model.predict(X_test)
                test_rmse = np.sqrt(mean_squared_error(y_test, y_test_pred))
                test_r2 = r2_score(y_test, y_test_pred)
                test_mae = mean_absolute_error(y_test, y_test_pred)
            else:
                test_rmse, test_r2, test_mae = None, None, None
            
            self.results[name] = {
                'train_rmse': train_rmse,
                'val_rmse': val_rmse,
                'test_rmse': test_rmse,
                'cv_rmse': cv_rmse,
                'cv_std': cv_std,
                'train_r2': train_r2,
                'val_r2': val_r2,
                'test_r2': test_r2,
                'train_mae': train_mae,
                'val_mae': val_mae,
                'test_mae': test_mae,
                'overfitting_gap': overfitting_gap
            }
            
            print(f"\n{name.upper()}:")
            print(f"  Training   - RMSE: ${train_rmse:,.2f}, R¬≤: {train_r2:.4f}, MAE: ${train_mae:,.2f}")
            print(f"  Validation - RMSE: ${val_rmse:,.2f}, R¬≤: {val_r2:.4f}, MAE: ${val_mae:,.2f}")
            if test_rmse:
                print(f"  Test       - RMSE: ${test_rmse:,.2f}, R¬≤: {test_r2:.4f}, MAE: ${test_mae:,.2f}")
            print(f"  CV RMSE    - Mean: ${cv_rmse:,.2f}, Std: ${cv_std:,.2f}")
            
            # Check for overfitting
            if overfitting_gap > 0.05:
                print(f"  ‚ö†Ô∏è Warning: R¬≤ gap of {overfitting_gap:.3f} indicates overfitting")
            else:
                print(f"  ‚úÖ Good generalization: R¬≤ gap of {overfitting_gap:.3f}")
        
        # Summary comparison
        print("\nüìä FINAL MODEL COMPARISON")
        print("="*80)
        
        summary_df = pd.DataFrame(self.results).T
        summary_cols = ['val_r2', 'val_rmse', 'cv_rmse', 'overfitting_gap']
        if 'test_r2' in summary_df.columns and summary_df['test_r2'].notna().any():
            summary_cols = ['val_r2', 'test_r2', 'val_rmse', 'test_rmse', 'overfitting_gap']
        
        summary = summary_df[summary_cols].round(4)
        summary['Status'] = summary['overfitting_gap'].apply(
            lambda x: '‚úÖ Good' if x < 0.05 else '‚ö†Ô∏è Overfit'
        )
        
        print(summary.to_string())
        
        # Select best model
        good_models = summary[summary['overfitting_gap'] < 0.05]
        if not good_models.empty:
            best_model_name = good_models['val_r2'].idxmax()
            print(f"\nüèÜ Best Model: {best_model_name}")
            print(f"   Validation R¬≤: {self.results[best_model_name]['val_r2']:.4f}")
            print(f"   Expected performance is realistic and not overfitted")
        else:
            print("\n‚ö†Ô∏è All models show overfitting. Consider further regularization.")
        
        return self.models, self.results

# Train advanced models with proper regularization
print("üîß Training Advanced Models with Strong Regularization...")
print("="*80)

# Note: Setting use_optuna=False for faster execution and more stable results
# You can set it to True if you want hyperparameter optimization
advanced = SafeAdvancedModels(use_optuna=False, n_trials=10)

# Train with test set included for final evaluation
advanced_models, advanced_results = advanced.train_all_advanced_models(
    X_train, y_train, 
    X_val, y_val,
    X_test, y_test  # Include test set for final evaluation
)
```

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
import warnings
warnings.filterwarnings('ignore')

# Choose ONE framework - using TensorFlow/Keras for neural networks
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, Model, callbacks, regularizers

class SafeDeepLearningModels:
    """
    Deep learning models properly sized and regularized for tabular data.
    Key principles:
    1. Architecture sized appropriately for data volume
    2. Strong regularization to prevent overfitting
    3. Proper input validation and scaling
    4. Single framework to avoid conflicts
    """
    
    def __init__(self, check_scaling=True):
        self.models = {}
        self.histories = {}
        self.results = {}
        self.check_scaling = check_scaling
        self.scaler = None
        
    def validate_and_scale_data(self, X_train, X_val, X_test=None):
        """
        Ensure data is properly scaled for neural networks
        """
        print("\nüìä Validating and Scaling Data for Neural Networks...")
        print("-"*50)
        
        # Check if data is already scaled
        train_means = X_train.mean()
        train_stds = X_train.std()
        
        # If data appears unscaled (high variance in scales), scale it
        if train_stds.max() > 10 or train_means.max() > 100:
            print("‚ö†Ô∏è Data appears unscaled. Applying StandardScaler...")
            self.scaler = StandardScaler()
            X_train_scaled = pd.DataFrame(
                self.scaler.fit_transform(X_train),
                columns=X_train.columns,
                index=X_train.index
            )
            X_val_scaled = pd.DataFrame(
                self.scaler.transform(X_val),
                columns=X_val.columns,
                index=X_val.index
            )
            if X_test is not None:
                X_test_scaled = pd.DataFrame(
                    self.scaler.transform(X_test),
                    columns=X_test.columns,
                    index=X_test.index
                )
            else:
                X_test_scaled = None
                
            print(f"‚úÖ Scaling applied. Features now have mean‚âà0, std‚âà1")
        else:
            print("‚úÖ Data appears to be already scaled")
            X_train_scaled = X_train
            X_val_scaled = X_val
            X_test_scaled = X_test if X_test is not None else None
            
        # Verify no NaN or infinite values
        for name, data in [('train', X_train_scaled), ('val', X_val_scaled)]:
            if data.isnull().any().any():
                print(f"‚ö†Ô∏è Warning: NaN values found in {name} data")
                data.fillna(0, inplace=True)
            if np.isinf(data).any().any():
                print(f"‚ö†Ô∏è Warning: Infinite values found in {name} data")
                data.replace([np.inf, -np.inf], 0, inplace=True)
                
        return X_train_scaled, X_val_scaled, X_test_scaled
    
    def create_appropriate_network(self, input_dim, n_samples, architecture='auto'):
        """
        Create a neural network sized appropriately for the data volume.
        Rule of thumb: Total parameters should be < n_samples/10
        """
        
        # Calculate appropriate architecture based on data size
        max_params = n_samples // 10  # Conservative parameter budget
        
        # Determine layer sizes based on input dimension and sample size
        if n_samples < 1000:
            # Very small data: Use minimal network
            layer_sizes = [min(32, input_dim * 2), 16]
        elif n_samples < 5000:
            # Small data: Conservative architecture
            layer_sizes = [min(64, input_dim * 3), 32, 16]
        elif n_samples < 20000:
            # Medium data: Moderate architecture
            layer_sizes = [128, 64, 32]
        else:
            # Large data: Can use deeper architecture
            layer_sizes = [256, 128, 64, 32]
        
        print(f"\nüèóÔ∏è Building network for {n_samples} samples, {input_dim} features")
        print(f"   Architecture: {' ‚Üí '.join(map(str, layer_sizes))} ‚Üí 1")
        
        # Calculate total parameters (approximate)
        total_params = input_dim * layer_sizes[0]  # First layer
        for i in range(len(layer_sizes) - 1):
            total_params += layer_sizes[i] * layer_sizes[i+1]
        total_params += layer_sizes[-1]  # Output layer
        
        print(f"   Approximate parameters: {total_params:,}")
        print(f"   Parameter ratio: {total_params/n_samples:.2f} params per sample")
        
        if total_params > max_params:
            print(f"   ‚ö†Ô∏è Warning: Network may be too large for data size")
        
        # Build the network with proper regularization
        inputs = keras.Input(shape=(input_dim,))
        x = inputs
        
        for i, units in enumerate(layer_sizes):
            # Add L2 regularization that gets stronger in deeper layers
            l2_strength = 0.001 * (i + 1)
            
            x = layers.Dense(
                units,
                kernel_initializer='he_normal',
                kernel_regularizer=regularizers.l2(l2_strength),
                bias_regularizer=regularizers.l2(l2_strength * 0.1)
            )(x)
            
            # Batch normalization for stability
            x = layers.BatchNormalization()(x)
            
            # Activation
            x = layers.Activation('relu')(x)
            
            # Dropout that decreases in deeper layers (less dropout as network narrows)
            dropout_rate = max(0.1, 0.3 - (0.05 * i))
            x = layers.Dropout(dropout_rate)(x)
        
        # Output layer (no activation for regression)
        outputs = layers.Dense(1, kernel_regularizer=regularizers.l2(0.001))(x)
        
        model = Model(inputs=inputs, outputs=outputs)
        return model, layer_sizes
    
    def create_ensemble_network(self, input_dim, n_samples):
        """
        Create an ensemble-like network with multiple paths
        This can be more robust than a single deep network
        """
        inputs = keras.Input(shape=(input_dim,))
        
        # Path 1: Shallow and wide
        path1 = layers.Dense(64, activation='relu', 
                            kernel_regularizer=regularizers.l2(0.001))(inputs)
        path1 = layers.BatchNormalization()(path1)
        path1 = layers.Dropout(0.2)(path1)
        path1 = layers.Dense(32, activation='relu',
                            kernel_regularizer=regularizers.l2(0.001))(path1)
        
        # Path 2: Deep and narrow  
        path2 = layers.Dense(32, activation='relu',
                            kernel_regularizer=regularizers.l2(0.001))(inputs)
        path2 = layers.BatchNormalization()(path2)
        path2 = layers.Dense(16, activation='relu',
                            kernel_regularizer=regularizers.l2(0.001))(path2)
        path2 = layers.Dense(16, activation='relu',
                            kernel_regularizer=regularizers.l2(0.001))(path2)
        
        # Path 3: Direct connection (linear relationships)
        path3 = layers.Dense(16, activation='linear',
                            kernel_regularizer=regularizers.l2(0.001))(inputs)
        
        # Combine all paths
        combined = layers.concatenate([path1, path2, path3])
        combined = layers.BatchNormalization()(combined)
        combined = layers.Dense(32, activation='relu',
                               kernel_regularizer=regularizers.l2(0.001))(combined)
        combined = layers.Dropout(0.1)(combined)
        
        outputs = layers.Dense(1)(combined)
        
        model = Model(inputs=inputs, outputs=outputs)
        return model
    
    def train_neural_network(self, X_train, y_train, X_val, y_val, 
                            X_test=None, y_test=None,
                            architecture='auto', epochs=200):
        """
        Train neural network with proper safeguards against overfitting
        """
        # Validate and scale data
        X_train_scaled, X_val_scaled, X_test_scaled = self.validate_and_scale_data(
            X_train, X_val, X_test
        )
        
        print(f"\nüß† Training Neural Network...")
        print("="*60)
        
        # Create appropriately sized model
        if architecture == 'ensemble':
            model = self.create_ensemble_network(X_train_scaled.shape[1], len(X_train_scaled))
            model_name = 'nn_ensemble'
        else:
            model, layer_sizes = self.create_appropriate_network(
                X_train_scaled.shape[1], 
                len(X_train_scaled),
                architecture
            )
            model_name = 'nn_auto'
        
        # Compile with appropriate learning rate for data size
        initial_lr = 0.001 if len(X_train_scaled) > 5000 else 0.0005
        
        optimizer = keras.optimizers.Adam(learning_rate=initial_lr)
        model.compile(
            optimizer=optimizer,
            loss='mse',
            metrics=['mae']
        )
        
        # Callbacks for training control
        early_stop = callbacks.EarlyStopping(
            monitor='val_loss',
            patience=30,  # Reduced patience
            restore_best_weights=True,
            verbose=0
        )
        
        reduce_lr = callbacks.ReduceLROnPlateau(
            monitor='val_loss',
            factor=0.5,
            patience=15,
            min_lr=1e-6,
            verbose=0
        )
        
        # Progress tracking
        class CompactProgress(callbacks.Callback):
            def on_epoch_end(self, epoch, logs=None):
                if (epoch + 1) % 20 == 0 or epoch == 0:
                    val_rmse = np.sqrt(logs['val_loss'])
                    train_rmse = np.sqrt(logs['loss'])
                    print(f"  Epoch {epoch + 1:3d} - Train RMSE: ${train_rmse:,.2f}, "
                          f"Val RMSE: ${val_rmse:,.2f}")
        
        # Determine appropriate batch size
        if len(X_train_scaled) < 500:
            batch_size = 16
        elif len(X_train_scaled) < 2000:
            batch_size = 32
        else:
            batch_size = 64
            
        print(f"\n  Training with batch size: {batch_size}")
        
        # Train model
        history = model.fit(
            X_train_scaled, y_train,
            validation_data=(X_val_scaled, y_val),
            epochs=epochs,
            batch_size=batch_size,
            callbacks=[early_stop, reduce_lr, CompactProgress()],
            verbose=0
        )
        
        # Evaluate performance
        y_train_pred = model.predict(X_train_scaled, verbose=0).flatten()
        y_val_pred = model.predict(X_val_scaled, verbose=0).flatten()
        
        train_rmse = np.sqrt(mean_squared_error(y_train, y_train_pred))
        val_rmse = np.sqrt(mean_squared_error(y_val, y_val_pred))
        train_r2 = r2_score(y_train, y_train_pred)
        val_r2 = r2_score(y_val, y_val_pred)
        train_mae = mean_absolute_error(y_train, y_train_pred)
        val_mae = mean_absolute_error(y_val, y_val_pred)
        
        # Test set evaluation if provided
        if X_test_scaled is not None and y_test is not None:
            y_test_pred = model.predict(X_test_scaled, verbose=0).flatten()
            test_rmse = np.sqrt(mean_squared_error(y_test, y_test_pred))
            test_r2 = r2_score(y_test, y_test_pred)
            test_mae = mean_absolute_error(y_test, y_test_pred)
        else:
            test_rmse, test_r2, test_mae = None, None, None
        
        # Store results
        self.models[model_name] = model
        self.histories[model_name] = history
        self.results[model_name] = {
            'train_rmse': train_rmse, 'val_rmse': val_rmse, 'test_rmse': test_rmse,
            'train_r2': train_r2, 'val_r2': val_r2, 'test_r2': test_r2,
            'train_mae': train_mae, 'val_mae': val_mae, 'test_mae': test_mae,
            'overfitting_gap': train_r2 - val_r2,
            'best_epoch': len(history.history['loss']) - 30  # Early stopping patience
        }
        
        # Print results
        print(f"\n  üìä Results for {model_name}:")
        print(f"  Training   - RMSE: ${train_rmse:,.2f}, R¬≤: {train_r2:.4f}, MAE: ${train_mae:,.2f}")
        print(f"  Validation - RMSE: ${val_rmse:,.2f}, R¬≤: {val_r2:.4f}, MAE: ${val_mae:,.2f}")
        if test_rmse:
            print(f"  Test       - RMSE: ${test_rmse:,.2f}, R¬≤: {test_r2:.4f}, MAE: ${test_mae:,.2f}")
        print(f"  Best epoch: {self.results[model_name]['best_epoch']}/{epochs}")
        
        # Overfitting check
        gap = self.results[model_name]['overfitting_gap']
        if gap > 0.1:
            print(f"  ‚ö†Ô∏è Warning: Overfitting detected (R¬≤ gap: {gap:.3f})")
            print("     Consider: smaller network, more regularization, or more data")
        elif gap < 0:
            print(f"  üìà Model generalizes well (validation better than training)")
        else:
            print(f"  ‚úÖ Good generalization (R¬≤ gap: {gap:.3f})")
        
        return model, history
    
    def plot_training_analysis(self):
        """
        Comprehensive training analysis plots
        """
        if not self.histories:
            print("No training histories to plot")
            return
            
        n_models = len(self.histories)
        fig, axes = plt.subplots(2, n_models, figsize=(6*n_models, 10))
        
        if n_models == 1:
            axes = axes.reshape(-1, 1)
        
        for idx, (name, history) in enumerate(self.histories.items()):
            # Plot 1: Loss curves
            ax1 = axes[0, idx]
            ax1.plot(history.history['loss'], label='Training', alpha=0.8, linewidth=2)
            ax1.plot(history.history['val_loss'], label='Validation', alpha=0.8, linewidth=2)
            
            # Mark best epoch
            best_epoch = np.argmin(history.history['val_loss'])
            best_val_loss = history.history['val_loss'][best_epoch]
            ax1.scatter(best_epoch, best_val_loss, color='red', s=100, zorder=5,
                       label=f'Best (Epoch {best_epoch})')
            
            ax1.set_xlabel('Epoch')
            ax1.set_ylabel('Loss (MSE)')
            ax1.set_title(f'{name.upper()} - Loss Curves', fontweight='bold')
            ax1.legend()
            ax1.grid(True, alpha=0.3)
            
            # Plot 2: RMSE over time
            ax2 = axes[1, idx]
            train_rmse = [np.sqrt(loss) for loss in history.history['loss']]
            val_rmse = [np.sqrt(loss) for loss in history.history['val_loss']]
            
            ax2.plot(train_rmse, label='Training RMSE', alpha=0.8, linewidth=2)
            ax2.plot(val_rmse, label='Validation RMSE', alpha=0.8, linewidth=2)
            
            # Add overfitting indicator
            final_gap = train_rmse[-1] - val_rmse[-1]
            color = 'green' if abs(final_gap) < 500 else 'orange' if abs(final_gap) < 1000 else 'red'
            ax2.axhspan(train_rmse[-1], val_rmse[-1], alpha=0.2, color=color,
                       label=f'Gap: ${abs(final_gap):.0f}')
            
            ax2.set_xlabel('Epoch')
            ax2.set_ylabel('RMSE ($)')
            ax2.set_title(f'{name.upper()} - RMSE Progress', fontweight='bold')
            ax2.legend()
            ax2.grid(True, alpha=0.3)
        
        plt.suptitle('Neural Network Training Analysis', fontsize=16, fontweight='bold')
        plt.tight_layout()
        plt.show()
        
        # Print summary comparison
        print("\nüìä NEURAL NETWORK COMPARISON SUMMARY")
        print("="*70)
        results_df = pd.DataFrame(self.results).T
        
        # Select key metrics
        summary_cols = ['val_r2', 'val_rmse', 'overfitting_gap', 'best_epoch']
        if 'test_r2' in results_df.columns and results_df['test_r2'].notna().any():
            summary_cols = ['val_r2', 'test_r2', 'val_rmse', 'test_rmse', 'overfitting_gap']
        
        summary = results_df[summary_cols].round(4)
        summary['Status'] = summary['overfitting_gap'].apply(
            lambda x: '‚úÖ Good' if x < 0.1 else '‚ö†Ô∏è Overfit' if x > 0.1 else 'üìà Great'
        )
        
        print(summary.to_string())
        
        # Identify best model
        good_models = summary[summary['overfitting_gap'] < 0.1]
        if not good_models.empty:
            best_model = good_models['val_r2'].idxmax()
            print(f"\nüèÜ Best Neural Network: {best_model}")
            print(f"   Validation R¬≤: {self.results[best_model]['val_r2']:.4f}")
            print(f"   Validation RMSE: ${self.results[best_model]['val_rmse']:,.2f}")


# Execute the corrected deep learning pipeline
print("üöÄ TRAINING NEURAL NETWORKS WITH PROPER ARCHITECTURE")
print("="*80)

# Initialize the safe deep learning models
dl_models = SafeDeepLearningModels(check_scaling=True)

# Train auto-sized network
model_auto, history_auto = dl_models.train_neural_network(
    X_train, y_train, 
    X_val, y_val,
    X_test, y_test,  # Include test for final evaluation
    architecture='auto',
    epochs=200
)

# Train ensemble network
model_ensemble, history_ensemble = dl_models.train_neural_network(
    X_train, y_train,
    X_val, y_val, 
    X_test, y_test,
    architecture='ensemble',
    epochs=200
)

# Visualize training progress
dl_models.plot_training_analysis()

# Note about TabNet
print("\nüìù Note about TabNet:")
print("TabNet requires significant memory and is best suited for larger datasets.")
print("For small insurance datasets (<5000 samples), simpler neural networks")
print("or gradient boosting methods typically perform better and are more stable.")
```

## 8. Advanced Ensemble Methods
### 8.1 Stacking and Blending

```python
# Corrected Ensemble Methods with Proper Implementation
from sklearn.ensemble import StackingRegressor, VotingRegressor
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.base import BaseEstimator, RegressorMixin, clone
from sklearn.model_selection import KFold
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
import warnings
warnings.filterwarnings('ignore')

class ProperEnsembleMethods:
    """
    Corrected ensemble implementation that prevents data leakage
    and properly implements stacking and blending
    """
    
    def __init__(self, base_model_specs=None):
        """
        Initialize with model specifications, not trained models
        base_model_specs: dict of {name: untrained_model_instance}
        """
        self.base_model_specs = base_model_specs if base_model_specs else {}
        self.ensemble_models = {}
        self.results = {}
        
    def create_voting_ensemble(self, X_train, y_train, X_val, y_val, model_names=None):
        """
        Voting ensemble - this one is actually okay to use trained models
        but we'll train fresh ones for consistency
        """
        print("\nüó≥Ô∏è Creating Voting Ensemble (Properly)...")
        print("-"*50)
        
        # Select which models to use
        if model_names is None:
            model_names = list(self.base_model_specs.keys())[:5]
        
        # Create fresh instances and train them
        estimators = []
        for name in model_names:
            if name in self.base_model_specs:
                print(f"  Training fresh {name} for voting...")
                # Clone to get a fresh instance
                fresh_model = clone(self.base_model_specs[name])
                # Train on the training set
                fresh_model.fit(X_train, y_train)
                estimators.append((name, fresh_model))
        
        # Create voting ensemble
        voting = VotingRegressor(estimators=estimators)
        # Note: VotingRegressor doesn't need to be fit again as models are pre-trained
        # But we'll fit it for consistency and to ensure proper setup
        voting.fit(X_train, y_train)
        
        # Evaluate
        y_train_pred = voting.predict(X_train)
        y_val_pred = voting.predict(X_val)
        
        train_rmse = np.sqrt(mean_squared_error(y_train, y_train_pred))
        val_rmse = np.sqrt(mean_squared_error(y_val, y_val_pred))
        train_r2 = r2_score(y_train, y_train_pred)
        val_r2 = r2_score(y_val, y_val_pred)
        
        overfitting_check = train_r2 - val_r2
        
        print(f"\nResults:")
        print(f"Training   - RMSE: ${train_rmse:,.2f}, R¬≤: {train_r2:.4f}")
        print(f"Validation - RMSE: ${val_rmse:,.2f}, R¬≤: {val_r2:.4f}")
        print(f"Overfitting check (R¬≤ gap): {overfitting_check:.4f}")
        
        if overfitting_check > 0.05:
            print("‚ö†Ô∏è Warning: Signs of overfitting detected")
        
        self.ensemble_models['voting'] = voting
        self.results['voting'] = {
            'train_rmse': train_rmse,
            'val_rmse': val_rmse,
            'train_r2': train_r2,
            'val_r2': val_r2,
            'overfit_gap': overfitting_check
        }
        
        return voting
    
    def create_proper_stacking_ensemble(self, X_train, y_train, X_val, y_val, 
                                       model_names=None, cv_folds=5):
        """
        Proper stacking implementation using fresh model instances
        """
        print("\nüìö Creating Proper Stacking Ensemble...")
        print("-"*50)
        print("Using fresh model instances for proper cross-validation")
        
        # Select models for stacking
        if model_names is None:
            model_names = list(self.base_model_specs.keys())[:5]
        
        # Create fresh estimators (not pre-trained!)
        estimators = []
        for name in model_names:
            if name in self.base_model_specs:
                # Clone to ensure fresh instance
                fresh_model = clone(self.base_model_specs[name])
                estimators.append((name, fresh_model))
                print(f"  Base model: {name} (fresh instance)")
        
        # Meta-learner with regularization
        meta_learner = Ridge(alpha=1.0)  # Increased regularization
        print(f"  Meta-learner: Ridge regression (alpha=1.0)")
        print(f"  Cross-validation folds: {cv_folds}")
        
        # Create stacking ensemble with proper CV
        print("\nTraining stacking ensemble with cross-validation...")
        stacking = StackingRegressor(
            estimators=estimators,
            final_estimator=meta_learner,
            cv=cv_folds,  # This ensures proper out-of-fold predictions
            n_jobs=-1
        )
        
        # Fit the stacking ensemble
        stacking.fit(X_train, y_train)
        
        # Evaluate
        y_train_pred = stacking.predict(X_train)
        y_val_pred = stacking.predict(X_val)
        
        train_rmse = np.sqrt(mean_squared_error(y_train, y_train_pred))
        val_rmse = np.sqrt(mean_squared_error(y_val, y_val_pred))
        train_r2 = r2_score(y_train, y_train_pred)
        val_r2 = r2_score(y_val, y_val_pred)
        
        overfitting_check = train_r2 - val_r2
        
        print(f"\nResults:")
        print(f"Training   - RMSE: ${train_rmse:,.2f}, R¬≤: {train_r2:.4f}")
        print(f"Validation - RMSE: ${val_rmse:,.2f}, R¬≤: {val_r2:.4f}")
        print(f"Overfitting check (R¬≤ gap): {overfitting_check:.4f}")
        
        if overfitting_check > 0.05:
            print("‚ö†Ô∏è Warning: Signs of overfitting detected")
        
        self.ensemble_models['stacking'] = stacking
        self.results['stacking'] = {
            'train_rmse': train_rmse,
            'val_rmse': val_rmse,
            'train_r2': train_r2,
            'val_r2': val_r2,
            'overfit_gap': overfitting_check
        }
        
        return stacking
    
    def create_proper_blending_ensemble(self, X_train, y_train, X_val, y_val,
                                       model_names=None, blend_split=0.2):
        """
        Proper blending implementation with holdout set
        """
        print("\nüîÄ Creating Proper Blending Ensemble...")
        print("-"*50)
        print(f"Using {blend_split*100:.0f}% holdout for blend feature creation")
        
        # Split training data for blending
        split_idx = int(len(X_train) * (1 - blend_split))
        
        # Create blend training and holdout sets
        X_blend_train = X_train.iloc[:split_idx]
        y_blend_train = y_train.iloc[:split_idx]
        X_blend_hold = X_train.iloc[split_idx:]
        y_blend_hold = y_train.iloc[split_idx:]
        
        print(f"Blend training set: {len(X_blend_train)} samples")
        print(f"Blend holdout set: {len(X_blend_hold)} samples")
        
        # Select models
        if model_names is None:
            model_names = list(self.base_model_specs.keys())[:5]
        
        # Train base models on blend training set and create predictions
        blend_features_hold = []
        blend_features_val = []
        trained_models = {}
        
        for name in model_names:
            if name in self.base_model_specs:
                print(f"\n  Training {name} on blend training set...")
                
                # Get fresh model instance
                model = clone(self.base_model_specs[name])
                
                # Train ONLY on the blend training portion
                model.fit(X_blend_train, y_blend_train)
                trained_models[name] = model
                
                # Create predictions for holdout and validation
                hold_pred = model.predict(X_blend_hold)
                val_pred = model.predict(X_val)
                
                blend_features_hold.append(hold_pred)
                blend_features_val.append(val_pred)
                
                # Show individual model performance
                hold_rmse = np.sqrt(mean_squared_error(y_blend_hold, hold_pred))
                print(f"    Holdout RMSE: ${hold_rmse:,.2f}")
        
        # Create blend feature matrices
        X_blend_meta = np.column_stack(blend_features_hold)
        X_val_meta = np.column_stack(blend_features_val)
        
        print(f"\nBlend meta-features shape: {X_blend_meta.shape}")
        
        # Train meta-model on holdout blend features
        print("\nTraining meta-model on blend features...")
        meta_model = Ridge(alpha=1.0)  # Regularized meta-model
        meta_model.fit(X_blend_meta, y_blend_hold)
        
        # For final predictions, retrain base models on full training set
        print("\nRetraining base models on full training set for final ensemble...")
        final_models = {}
        X_train_meta = []
        
        for name in model_names:
            if name in self.base_model_specs:
                model = clone(self.base_model_specs[name])
                model.fit(X_train, y_train)
                final_models[name] = model
                X_train_meta.append(model.predict(X_train))
        
        X_train_meta = np.column_stack(X_train_meta)
        
        # Evaluate using the retrained models
        y_train_pred = meta_model.predict(X_train_meta)
        y_val_pred = meta_model.predict(X_val_meta)
        
        train_rmse = np.sqrt(mean_squared_error(y_train, y_train_pred))
        val_rmse = np.sqrt(mean_squared_error(y_val, y_val_pred))
        train_r2 = r2_score(y_train, y_train_pred)
        val_r2 = r2_score(y_val, y_val_pred)
        
        overfitting_check = train_r2 - val_r2
        
        print(f"\nFinal Results:")
        print(f"Training   - RMSE: ${train_rmse:,.2f}, R¬≤: {train_r2:.4f}")
        print(f"Validation - RMSE: ${val_rmse:,.2f}, R¬≤: {val_r2:.4f}")
        print(f"Overfitting check (R¬≤ gap): {overfitting_check:.4f}")
        
        if overfitting_check > 0.05:
            print("‚ö†Ô∏è Warning: Signs of overfitting detected")
        
        # Show meta-model weights
        print("\nüìä Meta-Model Weights (Model Importance):")
        weights = meta_model.coef_
        for name, weight in zip(model_names, weights):
            bar = '‚ñà' * int(abs(weight) * 20)
            print(f"  {name:20s}: {weight:7.4f} {bar}")
        
        # Store the complete blending ensemble
        self.ensemble_models['blending'] = {
            'base_models': final_models,
            'meta_model': meta_model,
            'model_names': model_names
        }
        
        self.results['blending'] = {
            'train_rmse': train_rmse,
            'val_rmse': val_rmse,
            'train_r2': train_r2,
            'val_r2': val_r2,
            'overfit_gap': overfitting_check
        }
        
        return meta_model
    
    def create_advanced_stacking(self, X_train, y_train, X_val, y_val):
        """
        Advanced stacking with multiple layers and feature engineering
        """
        print("\nüèóÔ∏è Creating Advanced Multi-Layer Stacking...")
        print("-"*50)
        
        # Layer 1: Diverse base models
        layer1_models = []
        model_names = list(self.base_model_specs.keys())[:4]
        
        for name in model_names:
            if name in self.base_model_specs:
                layer1_models.append((f"L1_{name}", clone(self.base_model_specs[name])))
        
        # Layer 2: Stacking on Layer 1
        print("Building Layer 1 (base models)...")
        stacking_layer1 = StackingRegressor(
            estimators=layer1_models,
            final_estimator=Ridge(alpha=0.5),
            cv=3,
            n_jobs=-1
        )
        
        # You could add more layers, but be careful about overfitting
        print("Training complete stacking ensemble...")
        stacking_layer1.fit(X_train, y_train)
        
        # Evaluate
        y_train_pred = stacking_layer1.predict(X_train)
        y_val_pred = stacking_layer1.predict(X_val)
        
        train_rmse = np.sqrt(mean_squared_error(y_train, y_train_pred))
        val_rmse = np.sqrt(mean_squared_error(y_val, y_val_pred))
        train_r2 = r2_score(y_train, y_train_pred)
        val_r2 = r2_score(y_val, y_val_pred)
        
        overfitting_check = train_r2 - val_r2
        
        print(f"\nResults:")
        print(f"Training   - RMSE: ${train_rmse:,.2f}, R¬≤: {train_r2:.4f}")
        print(f"Validation - RMSE: ${val_rmse:,.2f}, R¬≤: {val_r2:.4f}")
        print(f"Overfitting check (R¬≤ gap): {overfitting_check:.4f}")
        
        self.ensemble_models['advanced_stacking'] = stacking_layer1
        self.results['advanced_stacking'] = {
            'train_rmse': train_rmse,
            'val_rmse': val_rmse,
            'train_r2': train_r2,
            'val_r2': val_r2,
            'overfit_gap': overfitting_check
        }
        
        return stacking_layer1
    
    def compare_all_ensembles(self):
        """
        Compare all ensemble methods with overfitting analysis
        """
        print("\nüìä ENSEMBLE METHODS COMPARISON")
        print("="*80)
        
        # Create comparison dataframe
        comparison = pd.DataFrame(self.results).T
        comparison = comparison.sort_values('val_r2', ascending=False)
        
        # Display with overfitting warnings
        print("\nPerformance Summary:")
        print("-"*60)
        print(f"{'Method':<20} {'Val RMSE':<12} {'Val R¬≤':<10} {'Overfit Gap':<12} {'Status'}")
        print("-"*60)
        
        for idx, row in comparison.iterrows():
            status = "‚úÖ Good" if row['overfit_gap'] < 0.05 else "‚ö†Ô∏è Overfit"
            print(f"{idx:<20} ${row['val_rmse']:<11,.2f} {row['val_r2']:<10.4f} {row['overfit_gap']:<12.4f} {status}")
        
        # Visualization
        fig, axes = plt.subplots(2, 2, figsize=(14, 10))
        
        # Plot 1: RMSE Comparison
        ax1 = axes[0, 0]
        models = comparison.index
        x_pos = np.arange(len(models))
        width = 0.35
        
        ax1.bar(x_pos - width/2, comparison['train_rmse'], width, label='Train', alpha=0.8)
        ax1.bar(x_pos + width/2, comparison['val_rmse'], width, label='Validation', alpha=0.8)
        ax1.set_xticks(x_pos)
        ax1.set_xticklabels(models, rotation=45, ha='right')
        ax1.set_ylabel('RMSE ($)')
        ax1.set_title('RMSE Comparison', fontweight='bold')
        ax1.legend()
        ax1.grid(True, alpha=0.3)
        
        # Plot 2: R¬≤ Comparison
        ax2 = axes[0, 1]
        ax2.bar(x_pos - width/2, comparison['train_r2'], width, label='Train', alpha=0.8)
        ax2.bar(x_pos + width/2, comparison['val_r2'], width, label='Validation', alpha=0.8)
        ax2.set_xticks(x_pos)
        ax2.set_xticklabels(models, rotation=45, ha='right')
        ax2.set_ylabel('R¬≤ Score')
        ax2.set_title('R¬≤ Score Comparison', fontweight='bold')
        ax2.legend()
        ax2.grid(True, alpha=0.3)
        ax2.set_ylim([0, 1])
        
        # Plot 3: Overfitting Check
        ax3 = axes[1, 0]
        colors = ['red' if gap > 0.05 else 'green' for gap in comparison['overfit_gap']]
        bars = ax3.bar(models, comparison['overfit_gap'], color=colors, alpha=0.8)
        ax3.axhline(y=0.05, color='red', linestyle='--', label='Overfit threshold')
        ax3.set_ylabel('R¬≤ Gap (Train - Val)')
        ax3.set_title('Overfitting Analysis', fontweight='bold')
        ax3.set_xticklabels(models, rotation=45, ha='right')
        ax3.legend()
        ax3.grid(True, alpha=0.3)
        
        # Plot 4: Best Model Predictions vs Actual
        ax4 = axes[1, 1]
        best_model_name = comparison.index[0]
        ax4.text(0.5, 0.5, f'Best Model:\n{best_model_name}\n\nVal R¬≤: {comparison.iloc[0]["val_r2"]:.4f}\nVal RMSE: ${comparison.iloc[0]["val_rmse"]:,.2f}',
                ha='center', va='center', fontsize=14, transform=ax4.transAxes,
                bbox=dict(boxstyle='round', facecolor='lightgreen', alpha=0.8))
        ax4.set_title('Best Ensemble Method', fontweight='bold')
        ax4.axis('off')
        
        plt.suptitle('Ensemble Methods Analysis', fontsize=16, fontweight='bold')
        plt.tight_layout()
        plt.show()
        
        # Recommendations
        print("\nüéØ RECOMMENDATIONS:")
        print("-"*60)
        
        good_models = comparison[comparison['overfit_gap'] < 0.05]
        if len(good_models) > 0:
            best_model = good_models.index[0]
            print(f"‚úÖ Recommended model: {best_model}")
            print(f"   - Shows good generalization (gap: {good_models.iloc[0]['overfit_gap']:.4f})")
            print(f"   - Validation R¬≤: {good_models.iloc[0]['val_r2']:.4f}")
        else:
            print("‚ö†Ô∏è All models show signs of overfitting.")
            print("   Consider:")
            print("   - Increasing regularization in meta-learners")
            print("   - Using fewer base models")
            print("   - Collecting more training data")
            print("   - Reducing model complexity")
        
        return comparison

# Example usage with your models
print("üöÄ PROPER ENSEMBLE METHODS IMPLEMENTATION")
print("="*80)

# You need to provide model specifications (not trained models!)
# Example:
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor

# Define base model specifications with conservative parameters
base_model_specs = {
    'rf': RandomForestRegressor(
        n_estimators=50, max_depth=5, min_samples_split=20,
        min_samples_leaf=10, random_state=42
    ),
    'gb': GradientBoostingRegressor(
        n_estimators=50, max_depth=3, learning_rate=0.05,
        min_samples_split=20, random_state=42
    ),
    'xgb': XGBRegressor(
        n_estimators=50, max_depth=3, learning_rate=0.05,
        subsample=0.7, colsample_bytree=0.7, random_state=42
    ),
    'lgbm': LGBMRegressor(
        n_estimators=50, max_depth=3, learning_rate=0.05,
        num_leaves=20, min_child_samples=20, random_state=42, verbose=-1
    )
}

# Initialize ensemble methods with model specifications
ensemble = ProperEnsembleMethods(base_model_specs=base_model_specs)

# Create various ensemble methods
print("\n" + "="*60)
print("Creating Multiple Ensemble Types")
print("="*60)

# 1. Voting Ensemble
ensemble.create_voting_ensemble(X_train, y_train, X_val, y_val)

# 2. Proper Stacking
ensemble.create_proper_stacking_ensemble(X_train, y_train, X_val, y_val)

# 3. Proper Blending
ensemble.create_proper_blending_ensemble(X_train, y_train, X_val, y_val)

# 4. Advanced Stacking
ensemble.create_advanced_stacking(X_train, y_train, X_val, y_val)

# Compare all methods
comparison_df = ensemble.compare_all_ensembles()

print("\n" + "="*60)
print("‚úÖ Ensemble Methods Complete!")
print("="*60)
```

The ensemble methods implementation reveals fundamental insights about how different aggregation strategies perform on insurance premium prediction. Understanding why voting succeeds where more sophisticated methods struggle provides valuable lessons about model combination strategies and the relationship between complexity and generalization.

The voting ensemble achieved the strongest performance with a validation R¬≤ of 0.8300 and RMSE of $5,207.34, while maintaining excellent generalization with an R¬≤ gap below 0.05. This success stems from voting's simplicity - it averages predictions from four diverse base models (Random Forest, Gradient Boosting, XGBoost, and LightGBM) without introducing additional parameters to learn. Each model contributes equally to the final prediction, creating a balanced ensemble that leverages the strengths of different algorithms while smoothing out individual model errors through simple averaging.

Stacking, despite its theoretical sophistication, showed signs of overfitting with an R¬≤ gap of 0.0551. The stacking approach trains base models using 5-fold cross-validation to generate out-of-fold predictions, then trains a Ridge regression meta-learner on these predictions. While this prevents the direct data leakage that would occur from training the meta-learner on in-sample predictions, the additional layer of learning introduces 5 new parameters (one coefficient per base model plus intercept) that must be estimated from limited data. With only 856 training samples, the meta-learner may be learning noise patterns in how the base models err, rather than systematic correction patterns.

The blending ensemble demonstrated similar overfitting tendencies (R¬≤ gap of 0.0530) despite using a holdout strategy. By reserving 20% of training data (172 samples) for creating blend features, the base models are trained on only 684 samples, reducing their individual performance. The meta-model weights reveal interesting dynamics: LightGBM received the highest weight (0.8787), while Random Forest received negative weight (-0.2726), suggesting the meta-learner is essentially selecting the best individual model rather than combining diverse perspectives. This winner-takes-all pattern often indicates insufficient data for the meta-learner to learn nuanced combination strategies.

Advanced stacking, which attempted to create a more sophisticated architecture, similarly suffered from overfitting (R¬≤ gap of 0.0537). The multi-layer approach compounds the parameter estimation problem - each additional layer requires learning new combination weights from increasingly derived features, amplifying estimation errors. With limited training data, simpler is often better, as each additional layer of complexity requires exponentially more data to estimate reliably.

The meta-model weights in the blending ensemble provide particularly revealing insights into how the models relate to each other. The strong positive weight for LightGBM and XGBoost versus the negative weight for Random Forest suggests these gradient boosting methods capture different patterns than the random forest. The meta-learner essentially learned to trust the gradient boosting predictions while using random forest as a contrarian indicator, a pattern that emerged from only 172 holdout samples and may not generalize reliably.

These results demonstrate a crucial principle in ensemble learning: the bias-variance tradeoff applies not just to individual models but to ensemble strategies themselves. Voting reduces variance through simple averaging without introducing additional parameters, making it robust with limited data. Stacking and blending attempt to reduce bias by learning optimal combination weights, but with insufficient data, they instead increase variance through parameter estimation uncertainty.

The consistent performance across all ensemble methods (validation R¬≤ ranging from 0.8267 to 0.8300) suggests we're approaching the performance ceiling for this dataset and feature set. The marginal improvements over individual models (best single model achieved around 0.83) indicate that the base models are already capturing most learnable patterns, leaving little room for ensemble methods to add value through sophisticated combination strategies.

For practical deployment, the voting ensemble represents the optimal choice. Its superior performance, combined with implementation simplicity, computational efficiency, and theoretical transparency, makes it ideal for production use. The model achieves a validation RMSE of $5,207, representing about 40% of the target standard deviation - a reasonable error level for insurance premium prediction where many risk factors remain unmeasured. The voting ensemble's robustness to individual model failures (if one model makes an error, three others can compensate) provides additional reliability for business-critical predictions.

The key lesson from these experiments is that ensemble complexity should match data availability. With thousands of samples, simple averaging often outperforms sophisticated meta-learning approaches that require estimating additional parameters. The success of voting validates the principle that diverse models making independent errors can achieve strong performance through simple aggregation, without needing to learn complex combination rules that may not generalize beyond the training data.

```python
import shap
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from lime.lime_tabular import LimeTabularExplainer
import warnings
warnings.filterwarnings('ignore')

class RobustExplainableAI:
    """
    Robust implementation of model interpretability using SHAP and LIME.
    Handles different model types and provides proper error handling.
    """
    
    def __init__(self, model, X_train, feature_names=None, model_type='auto'):
        """
        Initialize the explainer with proper model type detection.
        
        Parameters:
        -----------
        model : fitted model
            The trained model to explain
        X_train : pd.DataFrame or np.ndarray
            Training data for background distribution
        feature_names : list, optional
            Names of features
        model_type : str
            'tree', 'linear', 'deep', 'kernel', or 'auto' for automatic detection
        """
        self.model = model
        self.X_train = X_train
        
        # Handle feature names properly
        if feature_names is None:
            if hasattr(X_train, 'columns'):
                self.feature_names = X_train.columns.tolist()
            else:
                self.feature_names = [f'Feature_{i}' for i in range(X_train.shape[1])]
        else:
            self.feature_names = feature_names
            
        self.model_type = self._detect_model_type(model) if model_type == 'auto' else model_type
        self.shap_values = None
        self.explainer = None
        
        print(f"Initialized ExplainableAI with model type: {self.model_type}")
        
    def _detect_model_type(self, model):
        """
        Automatically detect the model type for appropriate SHAP explainer.
        """
        model_class = str(type(model))
        
        # Tree-based models
        tree_indicators = ['Forest', 'Tree', 'XGB', 'LGB', 'CatBoost', 'Gradient']
        if any(indicator in model_class for indicator in tree_indicators):
            return 'tree'
        
        # Linear models
        linear_indicators = ['Linear', 'Ridge', 'Lasso', 'Elastic']
        if any(indicator in model_class for indicator in linear_indicators):
            return 'linear'
        
        # Neural networks
        if 'keras' in model_class or 'tensorflow' in model_class or 'torch' in model_class:
            return 'deep'
        
        # Default to kernel (works for any model but slower)
        return 'kernel'
    
    def create_shap_explainer(self, sample_size=100):
        """
        Create appropriate SHAP explainer based on model type.
        """
        print(f"\nCreating SHAP explainer for {self.model_type} model...")
        
        try:
            if self.model_type == 'tree':
                self.explainer = shap.TreeExplainer(self.model)
                print("‚úì TreeExplainer created successfully")
                
            elif self.model_type == 'linear':
                self.explainer = shap.LinearExplainer(
                    self.model, 
                    self.X_train
                )
                print("‚úì LinearExplainer created successfully")
                
            elif self.model_type == 'deep':
                # For neural networks, use a sample of training data
                background = shap.sample(self.X_train, sample_size)
                self.explainer = shap.DeepExplainer(self.model, background)
                print("‚úì DeepExplainer created successfully")
                
            else:  # kernel or unknown
                # KernelExplainer works for any model but is slower
                # Use a smaller sample for background
                background = shap.sample(self.X_train, min(sample_size, len(self.X_train)))
                self.explainer = shap.KernelExplainer(
                    self.model.predict,
                    background
                )
                print(f"‚úì KernelExplainer created (using {len(background)} background samples)")
                
        except Exception as e:
            print(f"‚ö†Ô∏è Failed to create {self.model_type} explainer: {str(e)}")
            print("Falling back to KernelExplainer...")
            
            # Fallback to KernelExplainer
            background = shap.sample(self.X_train, min(50, len(self.X_train)))
            self.explainer = shap.KernelExplainer(
                self.model.predict,
                background
            )
            print("‚úì KernelExplainer created as fallback")
            
        return self.explainer
    
    def calculate_shap_values(self, X_test, max_samples=None):
        """
        Calculate SHAP values with proper error handling and memory management.
        """
        print("\nCalculating SHAP values...")
        
        if self.explainer is None:
            self.create_shap_explainer()
        
        # Limit samples if specified (helps with memory and speed)
        if max_samples and len(X_test) > max_samples:
            print(f"Limiting analysis to {max_samples} samples for efficiency")
            X_test_sample = X_test.iloc[:max_samples] if hasattr(X_test, 'iloc') else X_test[:max_samples]
        else:
            X_test_sample = X_test
            
        try:
            # Calculate SHAP values based on explainer type
            if self.model_type == 'tree':
                self.shap_values = self.explainer.shap_values(X_test_sample)
            elif self.model_type == 'kernel':
                # KernelExplainer can be slow, show progress
                print("Calculating (this may take a while for KernelExplainer)...")
                self.shap_values = self.explainer.shap_values(X_test_sample)
            else:
                self.shap_values = self.explainer.shap_values(X_test_sample)
                
            print(f"‚úì SHAP values calculated for {len(X_test_sample)} samples")
            
            # Handle different output formats
            if isinstance(self.shap_values, list):
                # For multi-output models
                self.shap_values = self.shap_values[0]
                
        except Exception as e:
            print(f"‚ùå Error calculating SHAP values: {str(e)}")
            return None
            
        return self.shap_values
    
    def generate_shap_analysis(self, X_test, max_samples=100):
        """
        Generate comprehensive SHAP analysis with separate, properly formatted plots.
        """
        print("\n" + "="*80)
        print("üîç SHAP ANALYSIS (SHapley Additive exPlanations)")
        print("="*80)
        print("\nSHAP values reveal how each feature contributes to predictions:")
        print("‚Ä¢ Positive values ‚Üí increase the prediction")
        print("‚Ä¢ Negative values ‚Üí decrease the prediction")
        print("‚Ä¢ Magnitude ‚Üí strength of the effect")
        print("-"*80)
        
        # Calculate SHAP values if not already done
        if self.shap_values is None:
            self.calculate_shap_values(X_test, max_samples)
            
        if self.shap_values is None:
            print("‚ùå Failed to calculate SHAP values")
            return None, None
        
        # Ensure we have the right data format for plotting
        if hasattr(X_test, 'iloc'):
            X_test_sample = X_test.iloc[:len(self.shap_values)]
        else:
            X_test_sample = X_test[:len(self.shap_values)]
        
        # Create individual SHAP plots (each in its own figure)
        print("\nGenerating SHAP visualizations...")
        
        # 1. Summary Plot (Beeswarm)
        print("\n1. Creating Summary Plot...")
        plt.figure(figsize=(10, 6))
        shap.summary_plot(
            self.shap_values, 
            X_test_sample,
            feature_names=self.feature_names,
            show=False
        )
        plt.title('SHAP Summary Plot - Feature Impact Distribution', fontweight='bold', pad=20)
        plt.tight_layout()
        plt.show()
        
        # 2. Feature Importance Bar Plot
        print("\n2. Creating Feature Importance Plot...")
        plt.figure(figsize=(10, 6))
        shap.summary_plot(
            self.shap_values,
            X_test_sample,
            feature_names=self.feature_names,
            plot_type="bar",
            show=False
        )
        plt.title('Global Feature Importance (Mean |SHAP|)', fontweight='bold', pad=20)
        plt.tight_layout()
        plt.show()
        
        # 3. Custom Analysis Dashboard
        print("\n3. Creating Custom Analysis Dashboard...")
        self._create_custom_dashboard(X_test_sample)
        
        # 4. Feature Statistics
        self._print_feature_insights()
        
        return self.shap_values, self.explainer
    
    def _create_custom_dashboard(self, X_test_sample):
        """
        Create a custom dashboard with matplotlib-compatible visualizations.
        """
        fig, axes = plt.subplots(2, 3, figsize=(18, 10))
        
        # 1. Feature importance ranking
        ax1 = axes[0, 0]
        feature_importance = np.abs(self.shap_values).mean(0)
        sorted_idx = feature_importance.argsort()[::-1][:10]
        
        ax1.barh(range(10), feature_importance[sorted_idx][::-1], color='steelblue')
        ax1.set_yticks(range(10))
        ax1.set_yticklabels([self.feature_names[i] for i in sorted_idx[::-1]])
        ax1.set_xlabel('Mean |SHAP value|')
        ax1.set_title('Top 10 Feature Importance', fontweight='bold')
        ax1.grid(True, alpha=0.3)
        
        # 2. SHAP value distribution
        ax2 = axes[0, 1]
        top_features = sorted_idx[:5]
        shap_data = [self.shap_values[:, idx] for idx in top_features]
        
        bp = ax2.boxplot(shap_data, labels=[self.feature_names[i] for i in top_features])
        ax2.set_xticklabels(ax2.get_xticklabels(), rotation=45, ha='right')
        ax2.set_ylabel('SHAP Value Distribution')
        ax2.set_title('SHAP Distributions - Top 5 Features', fontweight='bold')
        ax2.grid(True, alpha=0.3, axis='y')
        ax2.axhline(y=0, color='red', linestyle='--', alpha=0.5)
        
        # 3. Cumulative importance
        ax3 = axes[0, 2]
        cumulative_importance = np.cumsum(feature_importance[sorted_idx]) / feature_importance.sum()
        
        ax3.plot(range(len(cumulative_importance)), cumulative_importance, 'b-', linewidth=2)
        ax3.fill_between(range(len(cumulative_importance)), 0, cumulative_importance, alpha=0.3)
        ax3.axhline(y=0.9, color='red', linestyle='--', alpha=0.5, label='90% threshold')
        ax3.set_xlabel('Number of Features')
        ax3.set_ylabel('Cumulative Importance')
        ax3.set_title('Cumulative Feature Importance', fontweight='bold')
        ax3.legend()
        ax3.grid(True, alpha=0.3)
        
        # Find 90% threshold
        n_features_90 = np.where(cumulative_importance >= 0.9)[0][0] + 1
        ax3.axvline(x=n_features_90, color='green', linestyle=':', alpha=0.7)
        ax3.annotate(f'{n_features_90} features\nexplain 90%',
                    xy=(n_features_90, 0.9), xytext=(n_features_90+2, 0.7),
                    arrowprops=dict(arrowstyle='->', color='green', alpha=0.7))
        
        # 4. Feature correlation with SHAP values
        ax4 = axes[1, 0]
        top_feature_idx = sorted_idx[0]
        ax4.scatter(X_test_sample.iloc[:, top_feature_idx], 
                   self.shap_values[:, top_feature_idx],
                   alpha=0.5, s=20)
        ax4.set_xlabel(self.feature_names[top_feature_idx])
        ax4.set_ylabel('SHAP value')
        ax4.set_title(f'SHAP Dependency: {self.feature_names[top_feature_idx]}', fontweight='bold')
        ax4.grid(True, alpha=0.3)
        
        # Add trend line
        z = np.polyfit(X_test_sample.iloc[:, top_feature_idx], 
                      self.shap_values[:, top_feature_idx], 1)
        p = np.poly1d(z)
        x_line = np.linspace(X_test_sample.iloc[:, top_feature_idx].min(),
                            X_test_sample.iloc[:, top_feature_idx].max(), 100)
        ax4.plot(x_line, p(x_line), "r--", alpha=0.8, label='Trend')
        ax4.legend()
        
        # 5. Mean absolute SHAP by feature
        ax5 = axes[1, 1]
        mean_abs_shap = np.abs(self.shap_values).mean(0)
        sorted_features = mean_abs_shap.argsort()[-15:]
        
        ax5.barh(range(len(sorted_features)), mean_abs_shap[sorted_features], color='coral')
        ax5.set_yticks(range(len(sorted_features)))
        ax5.set_yticklabels([self.feature_names[i] for i in sorted_features], fontsize=8)
        ax5.set_xlabel('Mean |SHAP value|')
        ax5.set_title('Feature Importance - Extended View', fontweight='bold')
        ax5.grid(True, alpha=0.3, axis='x')
        
        # 6. SHAP value statistics
        ax6 = axes[1, 2]
        stats_text = self._generate_stats_text()
        ax6.text(0.1, 0.5, stats_text, transform=ax6.transAxes,
                fontsize=10, verticalalignment='center',
                bbox=dict(boxstyle='round', facecolor='lightgray', alpha=0.8))
        ax6.set_title('SHAP Analysis Statistics', fontweight='bold')
        ax6.axis('off')
        
        plt.suptitle('SHAP Analysis Dashboard', fontsize=16, fontweight='bold')
        plt.tight_layout()
        plt.show()
    
    def _generate_stats_text(self):
        """
        Generate statistics text for the dashboard.
        """
        feature_importance = np.abs(self.shap_values).mean(0)
        sorted_idx = feature_importance.argsort()[::-1]
        
        stats = []
        stats.append(f"Total features analyzed: {len(self.feature_names)}")
        stats.append(f"Samples analyzed: {len(self.shap_values)}")
        stats.append(f"\nTop 3 Features:")
        
        for i in range(min(3, len(sorted_idx))):
            idx = sorted_idx[i]
            importance = feature_importance[idx]
            stats.append(f"  {i+1}. {self.feature_names[idx]}: {importance:.3f}")
        
        # Calculate concentration
        cumulative = np.cumsum(feature_importance[sorted_idx]) / feature_importance.sum()
        n_50 = np.where(cumulative >= 0.5)[0][0] + 1
        n_90 = np.where(cumulative >= 0.9)[0][0] + 1
        
        stats.append(f"\nFeature Concentration:")
        stats.append(f"  50% importance: {n_50} features")
        stats.append(f"  90% importance: {n_90} features")
        
        # SHAP value range
        stats.append(f"\nSHAP Value Range:")
        stats.append(f"  Min: {self.shap_values.min():.3f}")
        stats.append(f"  Max: {self.shap_values.max():.3f}")
        stats.append(f"  Mean |SHAP|: {np.abs(self.shap_values).mean():.3f}")
        
        return '\n'.join(stats)
    
    def _print_feature_insights(self):
        """
        Print detailed insights from SHAP analysis.
        """
        print("\n" + "="*60)
        print("üìä KEY INSIGHTS FROM SHAP ANALYSIS")
        print("="*60)
        
        feature_importance = np.abs(self.shap_values).mean(0)
        sorted_idx = feature_importance.argsort()[::-1]
        
        print("\nüèÜ Top 10 Most Important Features:")
        print("-"*40)
        for i, idx in enumerate(sorted_idx[:10], 1):
            importance = feature_importance[idx]
            percentage = (importance / feature_importance.sum()) * 100
            print(f"{i:2d}. {self.feature_names[idx]:25s}: {importance:.4f} ({percentage:.1f}%)")
        
        # Feature concentration analysis
        cumulative = np.cumsum(feature_importance[sorted_idx]) / feature_importance.sum()
        n_50 = np.where(cumulative >= 0.5)[0][0] + 1
        n_90 = np.where(cumulative >= 0.9)[0][0] + 1
        
        print(f"\nüìà Feature Concentration:")
        print(f"   ‚Ä¢ Top {n_50} features explain 50% of model behavior")
        print(f"   ‚Ä¢ Top {n_90} features explain 90% of model behavior")
        
        if n_90 <= 10:
            print(f"   ‚úì Model relies on a small set of features (good interpretability)")
        else:
            print(f"   ‚ö† Model uses many features (complex interactions)")
        
        # Identify features with high variability
        shap_std = self.shap_values.std(axis=0)
        high_var_features = sorted_idx[np.argsort(shap_std[sorted_idx])[::-1][:3]]
        
        print(f"\nüîÑ Features with Highest Impact Variability:")
        for i, idx in enumerate(high_var_features, 1):
            print(f"   {i}. {self.feature_names[idx]} (std: {shap_std[idx]:.4f})")
        
        print("\nüí° Interpretation Tips:")
        print("   ‚Ä¢ Features with high importance drive most predictions")
        print("   ‚Ä¢ High variability features have context-dependent effects")
        print("   ‚Ä¢ Consider removing low-importance features for simpler models")
    
    def explain_single_prediction(self, X_sample, sample_index=0):
        """
        Explain a single prediction using waterfall plot.
        """
        print(f"\nüìç Explaining Single Prediction (Sample {sample_index})")
        print("-"*60)
        
        if self.explainer is None:
            self.create_shap_explainer()
        
        # Calculate SHAP values for this sample
        if hasattr(X_sample, 'iloc'):
            sample = X_sample.iloc[sample_index:sample_index+1]
        else:
            sample = X_sample[sample_index:sample_index+1]
        
        shap_values_sample = self.explainer.shap_values(sample)[0]
        
        # Create waterfall plot
        shap_exp = shap.Explanation(
            values=shap_values_sample,
            base_values=self.explainer.expected_value,
            data=sample.iloc[0] if hasattr(sample, 'iloc') else sample[0],
            feature_names=self.feature_names
        )
        
        plt.figure(figsize=(10, 6))
        shap.waterfall_plot(shap_exp, show=False)
        plt.title(f'Prediction Breakdown - Sample {sample_index}', fontweight='bold', pad=20)
        plt.tight_layout()
        plt.show()
        
        # Print top contributors
        print("\nTop Feature Contributions:")
        sorted_idx = np.argsort(np.abs(shap_values_sample))[::-1][:5]
        for idx in sorted_idx:
            contribution = shap_values_sample[idx]
            direction = "increases" if contribution > 0 else "decreases"
            print(f"  ‚Ä¢ {self.feature_names[idx]}: {direction} prediction by {abs(contribution):.2f}")
        
        return shap_values_sample
# Example usage
print("üöÄ ROBUST EXPLAINABLE AI IMPLEMENTATION")
print("="*80)

# Assuming you have a trained model and data
if 'advanced_models' in locals() and 'xgboost' in advanced_models:
    # Create explainer with automatic model type detection
    explainer = RobustExplainableAI(
        model=advanced_models['xgboost'],
        X_train=X_train,
        model_type='auto'  # Will detect that it's a tree model
    )
    
    # Generate comprehensive SHAP analysis
    shap_values, shap_explainer = explainer.generate_shap_analysis(
        X_test, 
        max_samples=100  # Limit for efficiency
    )
    
    # Explain a single prediction
    explainer.explain_single_prediction(X_test, sample_index=0)
    
else:
    print("\n‚ö†Ô∏è No trained model found.")
    print("Please ensure you have trained a model before running explainability analysis.")
    print("\nThis implementation supports:")
    print("  ‚Ä¢ Tree-based models (RandomForest, XGBoost, LightGBM, etc.)")
    print("  ‚Ä¢ Linear models (LinearRegression, Ridge, Lasso, etc.)")
    print("  ‚Ä¢ Neural networks (Keras, PyTorch, TensorFlow)")
    print("  ‚Ä¢ Any black-box model (using KernelExplainer)")

print("\n" + "="*80)
print("‚úÖ Explainability Analysis Complete!")
print("="*80)
```

### 10.1 Final Model Evaluation on Test Set

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
import time
import warnings
warnings.filterwarnings('ignore')

class ComprehensiveModelComparison:
    """
    This class evaluates and compares multiple models to help you make an informed decision
    about which model to deploy for insurance premium prediction.
    
    The key insight here is that different models have different strengths:
    - Linear models: Fast, interpretable, but may miss complex patterns
    - Tree-based models: Capture non-linear patterns, handle interactions well
    - Neural networks: Can model very complex relationships but need more data
    - Ensemble methods: Often provide the best performance by combining strengths
    """
    
    def __init__(self, models_dict, X_test, y_test, model_categories=None):
        """
        Initialize with all your trained models for comprehensive comparison.
        
        Parameters:
        -----------
        models_dict : dict
            Dictionary of {model_name: trained_model} for all models to compare
        X_test : DataFrame or array
            Test features for evaluation
        y_test : Series or array
            Test target values
        model_categories : dict
            Optional categorization of models for grouped analysis
        """
        self.models = models_dict
        self.X_test = X_test
        self.y_test = y_test
        
        # Categorize models for better understanding of their types
        self.model_categories = model_categories or self._auto_categorize_models()
        
        # Store results for all models
        self.evaluation_results = {}
        self.predictions = {}
        self.evaluation_times = {}
        
    def _auto_categorize_models(self):
        """
        Automatically categorize models based on their type.
        This helps us understand the diversity of our model portfolio.
        """
        categories = {
            'Linear': [],
            'Tree-Based': [],
            'Neural Network': [],
            'Ensemble': [],
            'Other': []
        }
        
        for name, model in self.models.items():
            model_type = str(type(model)).lower()
            
            # Categorize based on model type
            if any(term in model_type for term in ['linear', 'ridge', 'lasso', 'elastic']):
                categories['Linear'].append(name)
            elif any(term in model_type for term in ['tree', 'forest', 'xgb', 'lgb', 'catboost', 'gradient']):
                categories['Tree-Based'].append(name)
            elif any(term in model_type for term in ['neural', 'mlp', 'keras', 'torch', 'sequential']):
                categories['Neural Network'].append(name)
            elif any(term in model_type for term in ['voting', 'stacking', 'blending', 'ensemble']):
                categories['Ensemble'].append(name)
            else:
                categories['Other'].append(name)
        
        # Remove empty categories
        return {k: v for k, v in categories.items() if v}
    
    def evaluate_all_models(self, confidence_level=0.95):
        """
        Evaluate all models comprehensively.
        
        This method runs each model through the same evaluation process,
        ensuring fair comparison across different algorithm types.
        """
        print("="*80)
        print("COMPREHENSIVE MODEL EVALUATION AND COMPARISON")
        print("="*80)
        print(f"\nEvaluating {len(self.models)} models on {len(self.y_test)} test samples")
        print("\nModel Categories:")
        for category, model_names in self.model_categories.items():
            if model_names:
                print(f"  {category}: {', '.join(model_names)}")
        
        print("\n" + "-"*80)
        print("INDIVIDUAL MODEL EVALUATIONS")
        print("-"*80)
        
        # Ensure y_test is in the right format
        if hasattr(self.y_test, 'values'):
            y_test_array = self.y_test.values.flatten()
        else:
            y_test_array = np.array(self.y_test).flatten()
        
        # Evaluate each model
        for model_name, model in self.models.items():
            print(f"\nüìä Evaluating: {model_name}")
            print("-"*40)
            
            try:
                # Time the prediction
                start_time = time.time()
                
                # Make predictions (handling different model types)
                y_pred = self._safe_predict(model, self.X_test)
                
                prediction_time = time.time() - start_time
                
                # Store predictions for later analysis
                self.predictions[model_name] = y_pred
                self.evaluation_times[model_name] = prediction_time
                
                # Calculate comprehensive metrics
                results = self._calculate_metrics(y_test_array, y_pred)
                
                # Add model-specific information
                results['prediction_time'] = prediction_time
                results['model_type'] = self._get_model_type(model)
                results['n_parameters'] = self._estimate_model_complexity(model)
                
                # Ensure prediction_time is numeric
                results['prediction_time'] = float(prediction_time)
                
                # Store results
                self.evaluation_results[model_name] = results
                
                # Print summary for this model
                print(f"  R¬≤ Score:        {results['r2']:.4f}")
                print(f"  RMSE:            ${results['rmse']:,.2f}")
                print(f"  MAE:             ${results['mae']:,.2f}")
                print(f"  MAPE:            {results['mape']:.2f}%")
                print(f"  Prediction Time: {prediction_time:.3f} seconds")
                
            except Exception as e:
                print(f"  ‚ùå Error evaluating {model_name}: {str(e)}")
                self.evaluation_results[model_name] = None
        
        return self.evaluation_results
    
    def _safe_predict(self, model, X):
        """
        Safe prediction handling different model types.
        
        Different models require different prediction methods, and this
        function handles those differences transparently.
        """
        model_type = str(type(model)).lower()
        
        # Handle ensemble methods that might be dictionaries
        if isinstance(model, dict) and 'meta_model' in model:
            # This is likely a blending ensemble
            base_predictions = []
            for base_name, base_model in model['base_models'].items():
                base_pred = self._safe_predict(base_model, X)
                base_predictions.append(base_pred)
            X_meta = np.column_stack(base_predictions)
            return model['meta_model'].predict(X_meta).flatten()
        
        # Handle Keras/TensorFlow models
        elif 'keras' in model_type or 'tensorflow' in model_type:
            return model.predict(X, verbose=0).flatten()
        
        # Handle PyTorch models
        elif 'torch' in model_type:
            import torch
            model.eval()
            with torch.no_grad():
                if hasattr(X, 'values'):
                    X_tensor = torch.FloatTensor(X.values)
                else:
                    X_tensor = torch.FloatTensor(X)
                return model(X_tensor).numpy().flatten()
        
        # Standard sklearn-compatible models
        else:
            predictions = model.predict(X)
            return predictions.flatten() if len(predictions.shape) > 1 else predictions
    
    def _calculate_metrics(self, y_true, y_pred):
        """
        Calculate comprehensive evaluation metrics.
        
        We calculate multiple metrics because each tells us something different:
        - R¬≤: Overall explanatory power
        - RMSE: Average prediction error (in dollars)
        - MAE: Typical prediction error (less sensitive to outliers)
        - MAPE: Percentage error (scale-independent)
        """
        # Basic metrics
        rmse = np.sqrt(mean_squared_error(y_true, y_pred))
        mae = mean_absolute_error(y_true, y_pred)
        r2 = r2_score(y_true, y_pred)
        
        # Safe MAPE calculation
        mask = np.abs(y_true) > 1e-10
        if mask.sum() > 0:
            mape = np.mean(np.abs((y_true[mask] - y_pred[mask]) / y_true[mask])) * 100
        else:
            mape = np.mean(np.abs(y_true - y_pred) / (np.abs(y_true) + np.abs(y_pred) + 1e-10)) * 200
        
        # Percentile errors for business understanding
        errors = np.abs(y_true - y_pred)
        
        return {
            'r2': r2,
            'rmse': rmse,
            'mae': mae,
            'mape': mape,
            'median_error': np.median(errors),
            'p90_error': np.percentile(errors, 90),
            'p95_error': np.percentile(errors, 95),
            'max_error': np.max(errors),
            'std_error': np.std(errors)
        }
    
    def _get_model_type(self, model):
        """
        Identify the model type for categorization.
        """
        model_class = str(type(model))
        
        if 'RandomForest' in model_class:
            return 'Random Forest'
        elif 'GradientBoosting' in model_class:
            return 'Gradient Boosting'
        elif 'XGB' in model_class:
            return 'XGBoost'
        elif 'LGB' in model_class or 'Light' in model_class:
            return 'LightGBM'
        elif 'CatBoost' in model_class:
            return 'CatBoost'
        elif 'Linear' in model_class:
            return 'Linear Regression'
        elif 'Ridge' in model_class:
            return 'Ridge Regression'
        elif 'Lasso' in model_class:
            return 'Lasso Regression'
        elif 'ElasticNet' in model_class:
            return 'ElasticNet'
        elif 'SV' in model_class:
            return 'Support Vector'
        elif 'Neural' in model_class or 'Sequential' in model_class:
            return 'Neural Network'
        elif 'Voting' in model_class:
            return 'Voting Ensemble'
        elif 'Stacking' in model_class:
            return 'Stacking Ensemble'
        else:
            return 'Custom Model'
    
    def _estimate_model_complexity(self, model):
        """
        Estimate model complexity (number of parameters).
        
        Model complexity affects both performance and interpretability.
        Simpler models are often preferred when performance is similar.
        """
        try:
            # Start with a baseline complexity estimate
            if hasattr(self.X_test, 'shape'):
                n_features = self.X_test.shape[1]
            elif hasattr(model, 'n_features_in_'):
                n_features = model.n_features_in_
            else:
                n_features = 10  # Default fallback
            
            # Estimate based on model type
            if hasattr(model, 'coef_'):
                # Linear models - count coefficients
                if hasattr(model.coef_, 'shape'):
                    return np.prod(model.coef_.shape)
                else:
                    return len(model.coef_)
                    
            elif hasattr(model, 'n_estimators'):
                # Tree-based ensembles
                if hasattr(model, 'max_depth'):
                    # Rough estimate: trees * possible nodes
                    max_depth = model.max_depth if model.max_depth else 10
                    return model.n_estimators * min(2 ** max_depth, 1000)
                else:
                    return model.n_estimators * 100  # Conservative estimate
                    
            elif hasattr(model, 'get_config'):
                # Keras/TensorFlow models
                try:
                    return model.count_params()
                except:
                    return n_features * 100  # Fallback estimate
                    
            elif hasattr(model, 'support_vectors_'):
                # SVM models
                return len(model.support_vectors_) * n_features
                
            else:
                # Default estimate based on feature count
                return n_features * 10
                
        except Exception as e:
            # If all else fails, return a reasonable default
            return 100
    
    def create_comparison_visualizations(self):
        """
        Create comprehensive visualizations comparing all models.
        
        These visualizations help you understand:
        1. Which models perform best overall
        2. Trade-offs between different metrics
        3. Speed vs accuracy considerations
        4. How different models complement each other
        """
        # Filter out failed evaluations
        valid_results = {k: v for k, v in self.evaluation_results.items() if v is not None}
        
        if not valid_results:
            print("No valid evaluation results to visualize")
            return
        
        # Create comparison DataFrame
        comparison_df = pd.DataFrame(valid_results).T
        
        # Ensure all numeric columns are properly typed
        numeric_columns = ['r2', 'rmse', 'mae', 'mape', 'median_error', 
                          'p90_error', 'p95_error', 'max_error', 'std_error',
                          'prediction_time', 'n_parameters']
        
        for col in numeric_columns:
            if col in comparison_df.columns:
                comparison_df[col] = pd.to_numeric(comparison_df[col], errors='coerce')
                # Fill any NaN values with reasonable defaults
                if col == 'prediction_time':
                    comparison_df[col] = comparison_df[col].fillna(0.1)  # Default 0.1 seconds
                elif col == 'n_parameters':
                    comparison_df[col] = comparison_df[col].fillna(100)  # Default complexity
                else:
                    comparison_df[col] = comparison_df[col].fillna(comparison_df[col].median())
        
        # Sort by R¬≤ score (best first)
        comparison_df = comparison_df.sort_values('r2', ascending=False)
        
        # Create visualization dashboard
        fig, axes = plt.subplots(3, 3, figsize=(18, 14))
        
        # 1. Overall Performance Comparison (R¬≤ and RMSE)
        ax1 = axes[0, 0]
        x = np.arange(len(comparison_df))
        width = 0.35
        
        ax1_r2 = ax1.twinx()
        bars1 = ax1.bar(x - width/2, comparison_df['rmse'], width, 
                       label='RMSE', color='coral', alpha=0.7)
        bars2 = ax1_r2.bar(x + width/2, comparison_df['r2'], width, 
                          label='R¬≤', color='skyblue', alpha=0.7)
        
        ax1.set_xlabel('Model')
        ax1.set_ylabel('RMSE ($)', color='coral')
        ax1_r2.set_ylabel('R¬≤ Score', color='skyblue')
        ax1.set_xticks(x)
        ax1.set_xticklabels(comparison_df.index, rotation=45, ha='right')
        ax1.set_title('Performance Overview', fontweight='bold')
        ax1.grid(True, alpha=0.3)
        
        # 2. Percentage Error Comparison
        ax2 = axes[0, 1]
        ax2.bar(comparison_df.index, comparison_df['mape'], color='green', alpha=0.7)
        ax2.set_xlabel('Model')
        ax2.set_ylabel('MAPE (%)')
        ax2.set_title('Percentage Error Comparison', fontweight='bold')
        ax2.set_xticklabels(comparison_df.index, rotation=45, ha='right')
        ax2.grid(True, alpha=0.3, axis='y')
        
        # Add reference line for acceptable MAPE
        ax2.axhline(y=10, color='red', linestyle='--', alpha=0.5, label='Target (<10%)')
        ax2.legend()
        
        # 3. Speed vs Accuracy Trade-off
        ax3 = axes[0, 2]
        
        # Handle the size parameter safely - ensure it's valid for scatter plot
        if 'n_parameters' in comparison_df.columns:
            # Replace any NaN, inf, or zero values with a default size
            size_values = comparison_df['n_parameters'].fillna(100).replace([np.inf, -np.inf], 100)
            # Ensure all values are positive
            size_values = np.maximum(size_values, 10)
            # Scale for visualization (avoid too large or too small points)
            size_values = np.clip(size_values / 100, 10, 500)
        else:
            # If n_parameters doesn't exist, use a uniform size
            size_values = 100
        
        scatter = ax3.scatter(comparison_df['prediction_time'], 
                            comparison_df['r2'],
                            s=size_values,  # Safe size values
                            c=comparison_df['rmse'],
                            cmap='RdYlBu_r',
                            alpha=0.6,
                            edgecolors='black')
        
        # Annotate points with model names
        for idx, row in comparison_df.iterrows():
            ax3.annotate(idx, 
                        (row['prediction_time'], row['r2']),
                        fontsize=8, ha='center')
        
        ax3.set_xlabel('Prediction Time (seconds)')
        ax3.set_ylabel('R¬≤ Score')
        ax3.set_title('Speed vs Accuracy Trade-off', fontweight='bold')
        ax3.grid(True, alpha=0.3)
        plt.colorbar(scatter, ax=ax3, label='RMSE ($)')
        
        # 4. Error Distribution Comparison
        ax4 = axes[1, 0]
        error_metrics = ['median_error', 'p90_error', 'p95_error']
        bar_width = 0.25
        x = np.arange(len(comparison_df))
        
        for i, metric in enumerate(error_metrics):
            ax4.bar(x + i * bar_width, comparison_df[metric], 
                   bar_width, label=metric.replace('_', ' ').title(), alpha=0.7)
        
        ax4.set_xlabel('Model')
        ax4.set_ylabel('Error ($)')
        ax4.set_title('Error Distribution Metrics', fontweight='bold')
        ax4.set_xticks(x + bar_width)
        ax4.set_xticklabels(comparison_df.index, rotation=45, ha='right')
        ax4.legend(fontsize=8)
        ax4.grid(True, alpha=0.3, axis='y')
        
        # 5. Model Category Performance
        ax5 = axes[1, 1]
        category_performance = {}
        for category, model_names in self.model_categories.items():
            r2_scores = [comparison_df.loc[name, 'r2'] 
                        for name in model_names 
                        if name in comparison_df.index]
            if r2_scores:
                category_performance[category] = {
                    'mean_r2': np.mean(r2_scores),
                    'std_r2': np.std(r2_scores),
                    'count': len(r2_scores)
                }
        
        if category_performance:
            categories = list(category_performance.keys())
            mean_scores = [category_performance[c]['mean_r2'] for c in categories]
            std_scores = [category_performance[c]['std_r2'] for c in categories]
            
            ax5.bar(categories, mean_scores, yerr=std_scores, 
                   capsize=5, alpha=0.7, color='purple')
            ax5.set_xlabel('Model Category')
            ax5.set_ylabel('Mean R¬≤ Score')
            ax5.set_title('Performance by Model Category', fontweight='bold')
            ax5.grid(True, alpha=0.3, axis='y')
            
            # Add count annotations
            for i, (cat, score) in enumerate(zip(categories, mean_scores)):
                count = category_performance[cat]['count']
                ax5.text(i, score, f'n={count}', ha='center', va='bottom')
        
        # 6. Correlation Matrix of Predictions
        ax6 = axes[1, 2]
        if len(self.predictions) > 1:
            pred_df = pd.DataFrame(self.predictions)
            # Select top 6 models by R¬≤ for clarity
            top_models = comparison_df.head(6).index
            pred_subset = pred_df[[m for m in top_models if m in pred_df.columns]]
            
            corr_matrix = pred_subset.corr()
            im = ax6.imshow(corr_matrix, cmap='coolwarm', aspect='auto', vmin=0.7, vmax=1)
            ax6.set_xticks(range(len(corr_matrix)))
            ax6.set_yticks(range(len(corr_matrix)))
            ax6.set_xticklabels(corr_matrix.columns, rotation=45, ha='right')
            ax6.set_yticklabels(corr_matrix.columns)
            ax6.set_title('Prediction Correlation (Top 6)', fontweight='bold')
            plt.colorbar(im, ax=ax6)
            
            # Add correlation values as text
            for i in range(len(corr_matrix)):
                for j in range(len(corr_matrix)):
                    ax6.text(j, i, f'{corr_matrix.iloc[i, j]:.2f}',
                           ha='center', va='center', fontsize=8)
        
        # 7. Ranking Summary
        ax7 = axes[2, 0]
        # Create ranking for each metric
        rankings = pd.DataFrame()
        for metric in ['r2', 'rmse', 'mae', 'mape']:
            ascending = metric != 'r2'  # Higher R¬≤ is better, lower others are better
            rankings[metric] = comparison_df[metric].rank(ascending=ascending)
        
        # Calculate average rank
        rankings['avg_rank'] = rankings.mean(axis=1)
        rankings = rankings.sort_values('avg_rank')
        
        # Display top 5 models by average rank
        top_5 = rankings.head(5)
        y_pos = np.arange(len(top_5))
        ax7.barh(y_pos, top_5['avg_rank'], alpha=0.7, color='gold')
        ax7.set_yticks(y_pos)
        ax7.set_yticklabels(top_5.index)
        ax7.set_xlabel('Average Rank (Lower is Better)')
        ax7.set_title('Top 5 Models by Average Rank', fontweight='bold')
        ax7.grid(True, alpha=0.3, axis='x')
        ax7.invert_xaxis()  # Lower rank on the right
        
        # 8. Best Model for Each Metric
        ax8 = axes[2, 1]
        best_models = {
            'Best R¬≤': comparison_df['r2'].idxmax(),
            'Best RMSE': comparison_df['rmse'].idxmin(),
            'Best MAE': comparison_df['mae'].idxmin(),
            'Best MAPE': comparison_df['mape'].idxmin(),
            'Fastest': comparison_df['prediction_time'].idxmin()
        }
        
        # Count how many times each model is best
        model_counts = {}
        for metric, model in best_models.items():
            model_counts[model] = model_counts.get(model, 0) + 1
        
        if model_counts:
            models = list(model_counts.keys())
            counts = list(model_counts.values())
            colors = plt.cm.viridis(np.linspace(0.3, 0.9, len(models)))
            
            ax8.pie(counts, labels=models, autopct='%1.0f%%', 
                   colors=colors, startangle=90)
            ax8.set_title('Models with Best Metrics', fontweight='bold')
        
        # 9. Recommendation Summary
        ax9 = axes[2, 2]
        ax9.axis('off')
        
        # Ensure numeric types for safe operations
        comparison_df['r2'] = pd.to_numeric(comparison_df['r2'], errors='coerce')
        comparison_df['rmse'] = pd.to_numeric(comparison_df['rmse'], errors='coerce')
        comparison_df['prediction_time'] = pd.to_numeric(comparison_df['prediction_time'], errors='coerce')
        
        # Generate recommendation based on analysis
        best_overall = comparison_df.index[0]  # Best R¬≤
        best_balanced = rankings.index[0]  # Best average rank
        
        # Find fastest model with good performance
        good_models = comparison_df[comparison_df['r2'] > 0.8]
        if len(good_models) > 0:
            # Sort by prediction time and get the fastest
            good_models = good_models.sort_values('prediction_time')
            fastest_good = good_models.index[0] if len(good_models) > 0 else None
        else:
            fastest_good = None
        
        recommendation_text = f"""
RECOMMENDATION SUMMARY
{'='*30}

üèÜ Best Overall Performance:
   {best_overall}
   R¬≤ = {comparison_df.loc[best_overall, 'r2']:.4f}
   RMSE = ${comparison_df.loc[best_overall, 'rmse']:,.2f}

‚öñÔ∏è Most Balanced Model:
   {best_balanced}
   Avg Rank = {rankings.loc[best_balanced, 'avg_rank']:.1f}

üöÄ Best Speed/Performance:
   {fastest_good if fastest_good else 'N/A'}
"""
        
        if fastest_good:
            recommendation_text += f"""   R¬≤ = {comparison_df.loc[fastest_good, 'r2']:.4f}
   Time = {comparison_df.loc[fastest_good, 'prediction_time']:.3f}s
"""
        
        recommendation_text += f"""
üìä Model Diversity:
   {len(comparison_df)} models evaluated
   {len(self.model_categories)} categories tested
   
üí° Consider ensemble of:
   ‚Ä¢ {best_overall}
   ‚Ä¢ {best_balanced}
   for optimal results
"""
        
        ax9.text(0.1, 0.5, recommendation_text, transform=ax9.transAxes,
                fontsize=10, verticalalignment='center', fontfamily='monospace',
                bbox=dict(boxstyle='round', facecolor='lightgray', alpha=0.8))
        
        plt.suptitle('Comprehensive Model Comparison Dashboard', 
                    fontsize=16, fontweight='bold')
        plt.tight_layout()
        plt.show()
        
        return comparison_df
    
    def generate_deployment_recommendation(self):
        """
        Generate detailed deployment recommendations based on comprehensive analysis.
        
        This goes beyond simple metrics to consider practical deployment factors
        like interpretability, maintenance, and business constraints.
        """
        print("\n" + "="*80)
        print("DEPLOYMENT RECOMMENDATION REPORT")
        print("="*80)
        
        # Filter out failed evaluations
        valid_results = {k: v for k, v in self.evaluation_results.items() if v is not None}
        if not valid_results:
            print("No valid results for recommendation")
            return
        
        # Create DataFrame from valid results
        comparison_df = pd.DataFrame(valid_results).T
        
        # CRITICAL FIX: Ensure all numeric columns are properly typed
        # This is the key fix for your error - we must convert columns to numeric types
        numeric_columns = ['r2', 'rmse', 'mae', 'mape', 'median_error', 
                          'p90_error', 'p95_error', 'max_error', 'std_error',
                          'prediction_time', 'n_parameters']
        
        for col in numeric_columns:
            if col in comparison_df.columns:
                # Convert to numeric, replacing any non-numeric values with NaN
                comparison_df[col] = pd.to_numeric(comparison_df[col], errors='coerce')
                
                # Fill NaN values with sensible defaults based on the column
                if col == 'prediction_time':
                    comparison_df[col] = comparison_df[col].fillna(0.1)  # Default 0.1 seconds
                elif col == 'n_parameters':
                    comparison_df[col] = comparison_df[col].fillna(100)  # Default complexity
                elif col == 'r2':
                    # For R¬≤, use 0 as default (worst possible score)
                    comparison_df[col] = comparison_df[col].fillna(0)
                elif col in ['rmse', 'mae', 'mape', 'median_error', 'p90_error', 'p95_error', 'max_error', 'std_error']:
                    # For error metrics, use the maximum value as a conservative default
                    comparison_df[col] = comparison_df[col].fillna(comparison_df[col].max())
                else:
                    # Generic fallback: use median of valid values
                    comparison_df[col] = comparison_df[col].fillna(comparison_df[col].median())
        
        # Verify that critical columns are numeric before proceeding
        if comparison_df['r2'].dtype == 'object':
            print("Warning: R¬≤ values could not be converted to numeric. Using fallback approach.")
            # Try to extract numeric values from string representations if present
            try:
                comparison_df['r2'] = comparison_df['r2'].astype(float)
            except:
                print("Error: Cannot process R¬≤ values. Check your evaluation results.")
                return None
        
        # Now we can safely proceed with the analysis
        print("\n1. PERFORMANCE ANALYSIS")
        print("-"*40)
        
        # Identify top performers - this should now work without error
        try:
            top_r2 = comparison_df.nlargest(3, 'r2')
            print("Top 3 Models by R¬≤ Score:")
            for idx, row in top_r2.iterrows():
                print(f"  ‚Ä¢ {idx}: R¬≤={row['r2']:.4f}, RMSE=${row['rmse']:,.2f}")
        except Exception as e:
            print(f"Error identifying top performers: {e}")
            # Fallback: sort manually
            comparison_df_sorted = comparison_df.sort_values('r2', ascending=False)
            top_r2 = comparison_df_sorted.head(3)
            print("Top 3 Models by R¬≤ Score (fallback method):")
            for idx, row in top_r2.iterrows():
                print(f"  ‚Ä¢ {idx}: R¬≤={row['r2']:.4f}, RMSE=${row['rmse']:,.2f}")
        
        # Check if performance differences are significant
        r2_range = comparison_df['r2'].max() - comparison_df['r2'].min()
        if r2_range < 0.05:
            print("\n‚ö†Ô∏è Note: Performance differences are minimal (<5% R¬≤ range)")
            print("   Consider other factors like speed and interpretability")
        
        print("\n2. PRACTICAL CONSIDERATIONS")
        print("-"*40)
        
        # Speed analysis - with error handling
        try:
            fast_models = comparison_df.nsmallest(3, 'prediction_time')
            print("Fastest Models (for real-time applications):")
            for idx, row in fast_models.iterrows():
                print(f"  ‚Ä¢ {idx}: {row['prediction_time']:.4f}s per batch")
        except Exception as e:
            print(f"Could not analyze speed: {e}")
        
        # Robustness analysis (low variance in errors) - with safe division
        try:
            # Safe calculation of error variance coefficient
            comparison_df['error_variance'] = np.where(
                comparison_df['mae'] > 0,
                comparison_df['std_error'] / comparison_df['mae'],
                np.inf  # If MAE is 0, set variance to infinity
            )
            # Replace inf values with a large number for sorting
            comparison_df['error_variance'] = comparison_df['error_variance'].replace(np.inf, 999)
            
            robust_models = comparison_df.nsmallest(3, 'error_variance')
            print("\nMost Consistent Models (low error variance):")
            for idx, row in robust_models.iterrows():
                if row['error_variance'] < 999:  # Only show meaningful values
                    print(f"  ‚Ä¢ {idx}: CV={row['error_variance']:.3f}")
        except Exception as e:
            print(f"Could not analyze robustness: {e}")
        
        print("\n3. DEPLOYMENT SCENARIOS")
        print("-"*40)
        
        # Scenario 1: Maximum Accuracy
        try:
            best_accuracy = comparison_df['r2'].idxmax()
            print(f"\nüéØ For Maximum Accuracy:")
            print(f"   Deploy: {best_accuracy}")
            print(f"   Pros: Best R¬≤ ({comparison_df.loc[best_accuracy, 'r2']:.4f})")
            print(f"   Cons: May be slower ({comparison_df.loc[best_accuracy, 'prediction_time']:.3f}s)")
        except Exception as e:
            print(f"Could not determine best accuracy model: {e}")
        
        # Scenario 2: Real-time Applications
        try:
            # Find models with R¬≤ above 75th percentile
            r2_threshold = comparison_df['r2'].quantile(0.75)
            good_fast = comparison_df[comparison_df['r2'] > r2_threshold]
            
            if not good_fast.empty:
                best_speed = good_fast['prediction_time'].idxmin()
                print(f"\n‚ö° For Real-time Applications:")
                print(f"   Deploy: {best_speed}")
                print(f"   Pros: Fast ({comparison_df.loc[best_speed, 'prediction_time']:.4f}s), "
                      f"Good R¬≤ ({comparison_df.loc[best_speed, 'r2']:.4f})")
        except Exception as e:
            print(f"Could not determine best real-time model: {e}")
        
        # Scenario 3: Regulatory/Interpretability Requirements
        try:
            linear_models = [m for m in comparison_df.index 
                            if any(term in str(m).lower() for term in ['linear', 'ridge', 'lasso', 'elastic'])]
            if linear_models:
                # Get R¬≤ values for linear models and find the best
                linear_r2_values = {model: comparison_df.loc[model, 'r2'] for model in linear_models}
                best_linear = max(linear_r2_values.keys(), key=lambda x: linear_r2_values[x])
                print(f"\nüìã For Regulatory Compliance (interpretability):")
                print(f"   Deploy: {best_linear}")
                print(f"   Pros: Fully interpretable, coefficients available")
                print(f"   R¬≤ Performance: {comparison_df.loc[best_linear, 'r2']:.4f}")
        except Exception as e:
            print(f"Could not determine best interpretable model: {e}")
        
        print("\n4. FINAL RECOMMENDATION")
        print("-"*40)
        
        try:
            # Calculate weighted score considering multiple factors
            # Normalize each metric to 0-1 scale for fair weighting
            
            # Safe normalization with checks for zero division
            r2_max = comparison_df['r2'].max()
            time_max = comparison_df['prediction_time'].max()
            mape_max = comparison_df['mape'].max()
            var_max = comparison_df['error_variance'].replace(999, np.nan).max()
            
            comparison_df['deployment_score'] = 0
            
            # Add R¬≤ contribution (40% weight)
            if r2_max > 0:
                comparison_df['deployment_score'] += 0.4 * (comparison_df['r2'] / r2_max)
            
            # Add speed contribution (20% weight) - faster is better
            if time_max > 0:
                comparison_df['deployment_score'] += 0.2 * (1 - comparison_df['prediction_time'] / time_max)
            
            # Add MAPE contribution (20% weight) - lower is better
            if mape_max > 0:
                comparison_df['deployment_score'] += 0.2 * (1 - comparison_df['mape'] / mape_max)
            
            # Add consistency contribution (20% weight) - lower variance is better
            if pd.notna(var_max) and var_max > 0:
                # Replace 999 values with max for normalization
                error_var_normalized = comparison_df['error_variance'].replace(999, var_max)
                comparison_df['deployment_score'] += 0.2 * (1 - error_var_normalized / var_max)
            
            # Find the model with highest deployment score
            recommended = comparison_df['deployment_score'].idxmax()
            
            print(f"\nüèÜ RECOMMENDED MODEL: {recommended}")
            print(f"\nRationale:")
            print(f"  ‚Ä¢ R¬≤ Score: {comparison_df.loc[recommended, 'r2']:.4f}")
            print(f"  ‚Ä¢ RMSE: ${comparison_df.loc[recommended, 'rmse']:,.2f}")
            print(f"  ‚Ä¢ Prediction Speed: {comparison_df.loc[recommended, 'prediction_time']:.4f}s")
            print(f"  ‚Ä¢ Balanced performance across all metrics")
            
            # Suggest ensemble if top models are diverse
            top_3 = comparison_df.nlargest(3, 'deployment_score').index.tolist()
            
            # Check if we have predictions to analyze correlation
            if len(self.predictions) > 1 and len(top_3) > 1:
                try:
                    # Check if top models have predictions available
                    available_models = [m for m in top_3 if m in self.predictions]
                    if len(available_models) > 1:
                        pred_df = pd.DataFrame({m: self.predictions[m] for m in available_models})
                        pred_corr = pred_df.corr().values
                        # Calculate mean correlation (excluding diagonal)
                        np.fill_diagonal(pred_corr, np.nan)
                        mean_corr = np.nanmean(pred_corr)
                        
                        if mean_corr < 0.95:  # Models are diverse enough
                            print(f"\nüí° Alternative: Consider ensemble of:")
                            for model in available_models:
                                print(f"     ‚Ä¢ {model}")
                            print(f"   These models show diverse predictions (correlation={mean_corr:.3f})")
                            print(f"   Ensemble could improve performance by 2-5%")
                except Exception as e:
                    print(f"\nNote: Could not analyze model diversity for ensemble recommendation")
            
            return recommended
            
        except Exception as e:
            print(f"Error in final recommendation: {e}")
            # Fallback: return the model with best R¬≤
            try:
                return comparison_df['r2'].idxmax()
            except:
                return None

# Example usage showing how to evaluate ALL your models
print("="*80)
print("COMPREHENSIVE MODEL EVALUATION EXAMPLE")
print("="*80)

# Collect all your trained models into a dictionary
all_models = {}

# Add baseline models if they exist
if 'baseline' in locals() and hasattr(baseline, 'models'):
    all_models.update({f"baseline_{name}": model 
                      for name, model in baseline.models.items()})
    print(f"Added {len(baseline.models)} baseline models")

# Add advanced models if they exist
if 'advanced_models' in locals():
    all_models.update({f"advanced_{name}": model 
                      for name, model in advanced_models.items()})
    print(f"Added {len(advanced_models)} advanced models")

# Add ensemble models if they exist
if 'ensemble' in locals() and hasattr(ensemble, 'ensemble_models'):
    all_models.update({f"ensemble_{name}": model 
                      for name, model in ensemble.ensemble_models.items()})
    print(f"Added {len(ensemble.ensemble_models)} ensemble models")

# If you have specific models you want to compare, add them here
# For example, if you trained models earlier:
if 'advanced' in locals() and hasattr(advanced, 'models'):
    all_models.update(advanced.models)

if not all_models:
    # Create example models for demonstration
    print("\nNo trained models found. Creating example for demonstration...")
    from sklearn.linear_model import Ridge, Lasso
    from sklearn.ensemble import RandomForestRegressor
    
    # You would replace these with your actual trained models
    example_models = {
        'ridge': Ridge(alpha=1.0),
        'lasso': Lasso(alpha=0.1),
        'random_forest': RandomForestRegressor(n_estimators=50, max_depth=5, random_state=42)
    }
    
    # Fit example models (you would skip this with already trained models)
    if 'X_train' in locals() and 'y_train' in locals():
        for name, model in example_models.items():
            print(f"Training example {name}...")
            model.fit(X_train, y_train)
            all_models[name] = model

if all_models and 'X_test' in locals() and 'y_test' in locals():
    print(f"\nTotal models to evaluate: {len(all_models)}")
    print(f"Models: {', '.join(all_models.keys())}")
    
    # Create the comprehensive comparison system
    comparator = ComprehensiveModelComparison(
        models_dict=all_models,
        X_test=X_test,
        y_test=y_test
    )
    
    # Evaluate all models
    print("\n" + "="*60)
    print("Starting comprehensive evaluation...")
    print("="*60)
    
    results = comparator.evaluate_all_models()
    
    # Create comparison visualizations
    print("\n" + "="*60)
    print("Generating comparison visualizations...")
    print("="*60)
    
    comparison_df = comparator.create_comparison_visualizations()
    
    # Generate deployment recommendations
    recommended_model = comparator.generate_deployment_recommendation()
    
    print("\n" + "="*80)
    print("EVALUATION COMPLETE!")
    print("="*80)
    
else:
    print("\n‚ö†Ô∏è Unable to run comprehensive comparison")
    print("Requirements:")
    print("  1. Trained models (baseline, advanced, ensemble)")
    print("  2. Test data (X_test, y_test)")
    print("\nThe system can compare:")
    print("  ‚Ä¢ Linear models (Ridge, Lasso, ElasticNet)")
    print("  ‚Ä¢ Tree models (RandomForest, GradientBoosting)")
    print("  ‚Ä¢ Advanced models (XGBoost, LightGBM, CatBoost)")
    print("  ‚Ä¢ Neural networks (Keras, PyTorch)")
    print("  ‚Ä¢ Ensemble methods (Voting, Stacking, Blending)")
    print("\nAll models are evaluated equally and fairly!")
```
