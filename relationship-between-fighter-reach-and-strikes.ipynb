```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

df = pd.read_csv('/kaggle/input/ufc-datasets-1994-2025/UFC DATASETS/fighter_details.csv')

plt.figure(figsize=(8, 5))
sns.histplot(df['reach'], kde=True, color='purple')
plt.title('Distribution of Fighter Reach')
plt.xlabel('Reach (cm)')
plt.ylabel('Number of Fighters')
plt.show ()

```


Our first graph explores the distribution of reach across UFC fighters. We observe that most fighters have a reach clustered between 70 and 76 inches, with fewer athletes falling on either side of this range. This bell-like shape suggests there’s an ideal reach range most fighters fall into, likely reflecting common weight classes and body types. 

```python


import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

df = pd.read_csv('/kaggle/input/ufc-datasets-1994-2025/UFC DATASETS/fighter_details.csv')

plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='reach', y='str_acc', hue='stance', palette='Set2', alpha=0.7)
plt.title('Reach vs Strike Accuracy by Stance')
plt.xlabel('Reach (cm)')
plt.ylabel('Strike Accuracy (%)')
plt.legend(title='Stance', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```


To dive deeper, we charted Reach vs Strike Accuracy and included each fighter’s stance. Interestingly, while some fighters with longer reach do perform well, there’s no simple linear relationship, some fighters with shorter reach achieve high strike accuracy, while others with long reach perform modestly.

However, some clusters emerge, fighters with an Orthodox stance dominate the middle reach ranges, while Southpaw fighters are spread across the sides. This could hint at style differences in how reach is used.


```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

df = pd.read_csv('/kaggle/input/ufc-datasets-1994-2025/UFC DATASETS/fighter_details.csv')

df['win_group'] = pd.cut(df['wins'], bins=[0, 5, 10, 15, 20, 100], labels=['0-5', '6-10', '11-15', '16-20', '21+'])

plt.figure(figsize=(8, 6))
sns.scatterplot(
    data=df,
    x='td_avg_acc',
    y='td_def',    
    hue='stance',
    palette='Dark2',
    alpha=0.7
)
plt.title('Takedown Accuracy vs Defense by Stance')
plt.xlabel('Takedown Accuracy (%)')
plt.ylabel('Takedown Defense (%)')
plt.legend(title='Stance', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

This graph gives a 360-degree view of each fighter’s grappling effectiveness. It helps identify who is good at the wrestling aspect of MMA and who relies on other strategies to win fights. The takedown defence and takedown accuracy show a relationship between them, with most being around 40% Takedown Accuracy and 60% to 80% for the Takedown defence. This indicates that the Orthodox stance occupies a significant amount of space near these percentages. The 2nd largest would be southpaw, which also has a majority, but not as close to orthodox.
